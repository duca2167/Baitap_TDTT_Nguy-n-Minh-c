#### W9A2
Cho ma trận hai chiều nxm có các phần tử được sắp xếp trên dòng tăng dần và trên cột
tăng dần, hãy tìm số lớn thứ k (k < nxm) với thuật toán tốt nhất có thể

```python
n = int(input())
m = int(input())
matrix = []
for _ in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)
k = int(input())

def count_less_equal(matrix, target):
    n = len(matrix)
    m = len(matrix[0])
    count = 0
    row = 0
    col = m - 1
    while row < n and col >= 0:
        if matrix[row][col] <= target:
            count += n - row
            col -= 1
        else:
            row += 1
    return count

def find_kth_biggest(matrix, k):
    n = len(matrix)
    m = len(matrix[0])
    left = matrix[0][0]
    right = matrix[n - 1][m - 1]
    while left < right:
        mid = left + (right - left) // 2
        count = count_less_equal(matrix, mid)
        if count < k:
            left = mid + 1
        else:
            right = mid
    return left

print(find_kth_biggest(matrix, k))
    
```

#### W9A3
Cho một số nguyên dương, hãy thực hiện hoán đổi đúng 2 chữ số để được một số
nguyên nhỏ nhất có thể

```python
def hoan_doi_2_chu_so(so):
    lst = list(so)
    n = len(lst)
    for i in range(n - 1):
        min_val = min(lst[i+1:])
        if lst[i] > min_val:
            for j in range(n - 1, i, -1):
                if lst[j] == min_val:
                    lst[i], lst[j] = lst[j], lst[i]
                    return ''.join(lst)
    return ''.join(lst)

n = input()
print(hoan_doi_2_chu_so(n))
```

Đầu vào:
Dòng đầu tiên chứa 1 số nguyên Q là số lượng câu hỏi. Tương ứng với mỗi câu hỏi, đưa

)
Đầu ra:
Với mỗi câu hỏi, ghi ra kết quả tương ứng trên một dòng.

```python
def check_transformation(a, b):
    i = 0
    j = 0
    while i < len(a) and j < len(b):
        if a[i].upper() == b[j]:
            if a[i].isupper() and a[i] != b[j]:
                return "NO"
            j += 1
        elif a[i].isupper():
            return "NO"
        i += 1
    if j != len(b):
        return "NO"
    while i < len(a):
        if a[i].isupper():
            return "NO"
        i += 1
    return "YES"

q = int(input())
for _ in range(q):
    a = input()
    b = input()
    print(check_transformation(a, b))
```

[Hoán vị tiếp theo]
Cho một dãy số gồm n chữ số là một hoán vị của dãy 1,2,...,n. Viết chương trình in ra
hoán vị ở vị trí tiếp theo (danh sách các hoán vị được sắp xếp theo thứ tự từ điển) của
hoán vị đã cho.
Đầu vào từ bàn phím gồm 2 dòng:
- Dòng đầu tiên chứa số nguyên n
- Dòng tiếp theo chứa n số nguyên là một hoán vị của dãy 1,2,...,n (n≤105)
Đầu ra: In ra màn hình hoán vị tiếp theo của hoán vị đã cho.

```python
def hoan_vi_tiep_theo(n, arr):
    i = n - 2
    while i >= 0 and arr[i] >= arr[i + 1]:
        i -= 1
    if i < 0:
        return None
    j = n - 1
    while arr[j] <= arr[i]:
        j -= 1
    arr[i], arr[j] = arr[j], arr[i]
    left = i + 1
    right = n - 1
    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        left += 1
        right -= 1
    return arr
n = int(input())
arr = list(map(int, input().split()))
result = hoan_vi_tiep_theo(n, arr)
if result is None:
    print("NONE")
else:
    print(*result)
```

[CanPlaceFlowers]
Giả sử bạn là nhân viên của một công ty cây xanh, được giao cho chăm sóc một luống
đất. Luống đất đó được chia thành nhiều ô đất nhỏ có kích cỡ bằng nhau. Trước khi được
giao cho bạn chăm sóc, trên luống hoa đã được trồng sẵn một số cây xanh tán rộng. Một

vị trí còn trống trong luống hoa. Đây là những cây hoa hướng sáng và chúng không thể
sống được dưới bóng râm của cây khác, đặc biệt là cây tán rộng. Vì vậy, bạn không thể
trồng chúng tại ô bên cạnh những ô đã được trồng cây xanh tán rộng. Và nếu trồng những
cây hoa bên cạnh nhau, chúng sẽ tranh giành phân bón và nước và sau cùng, cả hai cây
sẽ cùng chết.
Viết hàm canPlaceFlowers(flowerbed, k) kiểm tra xem có thể trồng hết số hoa mới được
cấp vào trong luống đất của bạn hay không.
Hàm nhận đầu vào là một danh sách (list) flowerbed, đánh dấu những vị trí trong luống đất
đã được trồng cây xanh trước đó (giá trị 1 ứng với vị trí cây đã được trồng và 0 ứng với vị
trí đất trống) và số lượng hoa được phát thêm k (int). Hàm trả về True nếu có thể trồng hết
toàn bộ k cây hoa vào luống đất và False trong trường hợp ngược lại.

```python
def canPlaceFlowers(flowerbed, n):
    count = 0
    length = len(flowerbed)
    for i in range(length):
        if flowerbed[i] == 0:
            left_empty = (i == 0) or (flowerbed[i - 1] == 0)
            right_empty = (i == length - 1) or (flowerbed[i + 1] == 0)
            if left_empty and right_empty:
                flowerbed[i] = 1
                count += 1
    return count >= n
flowerbed = list(map(int, input().split()))
n = int(input())
print(canPlaceFlowers(flowerbed, n))
```

#### W9A8
[Phát hiện đường đi lặp lại của Rô-bốt]
Ban đầu, một con rô-bốt ở vị trí (0, 0) trên một lưới 2D.
Cho một chuỗi `moves` chứa các thao tác di chuyển của rô-bốt.
Các bước di chuyển được biểu diễn bởi các ký tự:
- 'R' (Phải)
- 'L' (Trái)
- 'U' (Lên)
- 'D' (Xuống)
Viết hàm `def is_path_crossing(moves)`:
Hàm nhận đầu vào là một chuỗi `moves`.
Hàm trả về `True` nếu trong quá trình di chuyển, rô-bốt đi vào một ô mà nó đã từng đi qua
trước đó.
Hàm trả về `False` nếu rô-bốt không bao giờ đi lặp lại (luôn đi vào ô mới).
Lưu ý: Vị trí ban đầu (0, 0) được tính là vị trí đầu tiên đã đi qua

```python
def is_path_crossing(moves):
    x, y = 0, 0
    visited = {(0, 0)}
    for move in moves:
        if move == 'U':
            y += 1
        elif move == 'D':
            y -= 1
        elif move == 'R':
            x += 1
        elif move == 'L':
            x -= 1
        if (x, y) in visited:
            return True
        visited.add((x, y))
    return False
moves = input()
print(is_path_crossing(moves))
```

Phần đọc input và gọi hàm (hàm main) đã có sẵn.

```python
def Bubblesort(nums):
    n = len(nums)
    for i in range(n):
        for j in range(0, n - i - 1):
            if nums[j] > nums[j + 1]:
                nums[j], nums[j + 1] = nums[j + 1], nums[j]
def Median(nums):
    Bubblesort(nums)
    n = len(nums)
    if n % 2 == 0:
        return (nums[n // 2 - 1] + nums[n // 2]) / 2
    else:
        return nums[n // 2]
nums = list(map(int, input().split()))
print(Median(nums))
```

#### W9A10
[Tìm bán kính lò sưởi nhỏ nhất]
Mùa đông đang đến! Bạn được cung cấp vị trí của các ngôi nhà và các lò sưởi trên một
đường thẳng.
Nhiệm vụ của bạn là tìm bán kính r nhỏ nhất cho các lò sưởi sao cho mọi ngôi nhà đều
nằm trong phạm vi sưởi ấm của ít nhất một lò sưởi.
Tất cả các lò sưởi đều có cùng một bán kính r.
Một ngôi nhà tại vị trí h được sưởi ấm bởi một lò sưởi tại vị trí t nếu |h - t| <= r.
Viết hàm `def findRadius(houses, heaters):`
Hàm nhận đầu vào là hai danh sách (list) các số nguyên:
- `houses`: chứa vị trí của các ngôi nhà.
- `heaters`: chứa vị trí của các lò sưởi.
Hàm trả về một số nguyên là bán kính r nhỏ nhất cần thiết.
Lưu ý:
- Số lượng nhà và lò sưởi là một số không âm và không vượt quá 25000.
- Vị trí của các ngôi nhà và lò sưởi là một số không âm và không vượt quá 10^9.

```python
def findRadius(houses, heaters):
    heaters.sort()
    max_radius = 0
    for house in houses:
        min_dist = float('inf')
        for heater in heaters:
            dist = abs(house - heater)
            if dist < min_dist:
                min_dist = dist
            elif heater > house:
                break
        if min_dist > max_radius:
            max_radius = min_dist
    return max_radius
houses = list(map(int, input().split()))
heaters = list(map(int, input().split()))
print(findRadius(houses, heaters))
```

```python
#CÁCH LÀM TỪ AI ( ĐƯA VÀO ĐÂY ĐỂ HỌC HỎI)
import bisect

def findRadius(houses, heaters):
    # Sắp xếp để dùng tìm kiếm nhị phân
    heaters.sort()
    max_radius = 0
    
    for house in houses:
        # Tìm vị trí thích hợp của ngôi nhà trong danh sách lò sưởi
        idx = bisect.bisect_left(heaters, house)
        
        # Khoảng cách đến lò sưởi bên trái
        dist_left = house - heaters[idx - 1] if idx > 0 else float('inf')
        
        # Khoảng cách đến lò sưởi bên phải
        dist_right = heaters[idx] - house if idx < len(heaters) else float('inf')
        
        # Khoảng cách gần nhất đến lò sưởi của ngôi nhà này
        min_dist = min(dist_left, dist_right)
        
        # Bán kính cần tìm là giá trị lớn nhất trong các min_dist
        max_radius = max(max_radius, min_dist)
        
    return max_radius
```

#### W9A11 Viết hàm factorial(n) tính giai thừa của n bằng đệ quy.

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)
n = int(input())
print(factorial(n))
```

#### W9A14
VIế́t hàm tìm số xuất hiện nhiều hơn x lần trong một dãy số (x là số nguyên dương nhập
từ bàn phím). Nếu không có số nào thì in ra "INVALID"

```python
def more_than_x(arr, x):
    if len(arr) == 0:
        return "INVALID"
    else:
        count = {}
        for i in arr:
            if i in count:
                count[i] += 1
            else:
                count[i] = 1
        for key, value in count.items():
            if value > x:
                return key
        return "INVALID"
x = int(input())
arr = list(map(int, input().split()))
print(more_than_x(arr, x))
```
