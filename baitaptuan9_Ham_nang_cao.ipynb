#### W9A2
Cho ma trận hai chiều nxm có các phần tử được sắp xếp trên dòng tăng dần và trên cột
tăng dần, hãy tìm số lớn thứ k (k < nxm) với thuật toán tốt nhất có thể

```python
n = int(input())
m = int(input())
matrix = []
for _ in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)
k = int(input())

def count_less_equal(matrix, target):
    n = len(matrix)
    m = len(matrix[0])
    count = 0
    row = 0
    col = m - 1
    while row < n and col >= 0:
        if matrix[row][col] <= target:
            count += n - row
            col -= 1
        else:
            row += 1
    return count

def find_kth_biggest(matrix, k):
    n = len(matrix)
    m = len(matrix[0])
    left = matrix[0][0]
    right = matrix[n - 1][m - 1]
    while left < right:
        mid = left + (right - left) // 2
        count = count_less_equal(matrix, mid)
        if count < k:
            left = mid + 1
        else:
            right = mid
    return left

print(find_kth_biggest(matrix, k))
    
```

#### W9A3
Cho một số nguyên dương, hãy thực hiện hoán đổi đúng 2 chữ số để được một số
nguyên nhỏ nhất có thể

```python
def hoan_doi_2_chu_so(so):
    lst = list(so)
    n = len(lst)
    for i in range(n - 1):
        min_val = min(lst[i+1:])
        if lst[i] > min_val:
            for j in range(n - 1, i, -1):
                if lst[j] == min_val:
                    lst[i], lst[j] = lst[j], lst[i]
                    return ''.join(lst)
    return ''.join(lst)

n = input()
print(hoan_doi_2_chu_so(n))
```

#### W9A4
Cho xâu a gồm các chữ cái thường và in hoa và xâu b chỉ gồm các chữ cái in hoa. Ta có
thể biến đổi xâu a theo 2 bước như sau:
- Chuyển một số chữ cái thường trong xâu a thành chữ cái in hoa (có thể không biến đổi
chữ cái nào)
- Xóa hết tất cả các chữ cái thường trong xâu a.
Cho 2 xâu a,b, xác đính xem có cách biến đổi từ xâu a sang xâu b không. Ghi ra "YES"
nếu có và "NO" nếu ngược lại.
Đầu vào:
Dòng đầu tiên chứa 1 số nguyên Q là số lượng câu hỏi. Tương ứng với mỗi câu hỏi, đưa
ra 2 xâu a,b trên 2 dòng khác nhau.(Q≤10,|a|,|b|≤1000
)
Đầu ra:
Với mỗi câu hỏi, ghi ra kết quả tương ứng trên một dòng.

```python
def check_transformation(a, b):
    i = 0
    j = 0
    while i < len(a) and j < len(b):
        if a[i].upper() == b[j]:
            if a[i].isupper() and a[i] != b[j]:
                return "NO"
            j += 1
        elif a[i].isupper():
            return "NO"
        i += 1
    if j != len(b):
        return "NO"
    while i < len(a):
        if a[i].isupper():
            return "NO"
        i += 1
    return "YES"

q = int(input())
for _ in range(q):
    a = input()
    b = input()
    print(check_transformation(a, b))
```

#### W9A5
Cho hai mảng số nguyên list1 và list2, viết hàm kiểm tra xem list1 có thuộc list2 hay không
(có phải là dãy con liên tiếp của list2 hay không)

```python
list1 = list(map(int, input().split()))
list2 = list(map(int, input().split()))
def list1_con_list2(list1, list2):
    n = len(list1)
    m = len(list2)
    if n == 0:
        return True
    for i in range(m - n + 1):
        if list2[i:i + n] == list1:
            return True
    return False
print(list1_con_list2(list1, list2))
```

#### W9A11 Viết hàm factorial(n) tính giai thừa của n bằng đệ quy.

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)
n = int(input())
print(factorial(n))
```

#### W9A12 Viết hàm sum_to_n(n) tính tổng từ 1 đến n bằng đệ quy.

```python
def sum_to_n(n):
    if n == 1:
        return 1
    return n+sum_to_n(n-1)
n = int(input())
print(sum_to_n(n))
```

#### W9A13
Viết hàm power(base, exp) tính base^exp bằng đệ quy chia để trị.
Yêu cầu:
- Không dùng vòng lặp
- Không dùng hàm có sẵn

```python
def power(base, exp):
    if exp == 0:
        return 1
    return base * power(base, exp - 1)
base = int(input())
exp = int(input())
print(power(base, exp))
```

#### W9A14
VIế́t hàm tìm số xuất hiện nhiều hơn x lần trong một dãy số (x là số nguyên dương nhập
từ bàn phím). Nếu không có số nào thì in ra "INVALID"

```python
def more_than_x(arr, x):
    if len(arr) == 0:
        return "INVALID"
    else:
        count = {}
        for i in arr:
            if i in count:
                count[i] += 1
            else:
                count[i] = 1
        for key, value in count.items():
            if value > x:
                return key
        return "INVALID"
x = int(input())
arr = list(map(int, input().split()))
print(more_than_x(arr, x))
```

#### W9A15
VIết haàm kiểm tra dãy số nguyên vào từ bàn phím có phải dãy tăng nghiêm ngặt (giảm
nghiêm ngặt) hay không?

```python
#TĂNG NGHIÊM NGẶT
def tang_nghiem_ngat(arr):
    tang = True
    n = len(arr)
    for i in range(1, n):
        if arr[i] <= arr[i-1]:
            tang = False
            return tang
    return tang

arr = list(map(int, input().split()))
print(tang_nghiem_ngat(arr))
```
