```python
a=float(input("Nhập chiều rộng a: "))
b=float(input("Nhập chiều dài b: "))
r=a/2
scircle=3.14*r**2
sleft=a*b-scircle
print("Diện tích phần còn lại là: ", sleft)
```

#### Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không, nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”

```python
a= float(input("Nhập số a: "))
b= float(input("Nhập số b: "))
c= float(input("Nhập số c: "))
if a+b>c and a+c>b and b+c>a:
    if a==b==c:
        print("Đây là một tam giác đều")
    elif a==b or a==c or b==c:
        print("Đây là một tam giác cân")
    else:
        print("Đây là một tam giác thường")
    p=(a+b+c)/2
    s=round((p*(p-a)*(p-b)*(p-c))**0.5, 1)
    print(f"Diện tích tam giác là: {s}")
else:
    print("Ba cạnh a, b, c không thể tạo thành một tam giác")
```

- Bậc 3: Từ 101 – 200 kWh: 2.380 đồng/kWh
- Bậc 4: Từ 201 – 300 kWh: 2.998 đồng/kWh
- Bậc 5: Từ 301 – 400 kWh: 3.350 đồng/kWh
- Bậc 6: Từ 401 trở đi kWh: 3.460 đồng/kWh
Tính thêm tiền VAT 8%. Chỉ lấy đến đơn vị đồng

```python

ten = input("Ten chu ho: ")
cu = int(input("Chi so thang truoc: "))
moi = int(input("Chi so thang nay: "))
kwh = moi - cu
tien_chua_vat = 0
if kwh <= 50:
    tien_chua_vat = kwh * 1984
elif kwh <= 100:
    tien_chua_vat = 50 * 1984 + (kwh - 50) * 2050
elif kwh <= 200:
    tien_chua_vat = 50 * 1984 + 50 * 2050 + (kwh - 100) * 2380
elif kwh <= 300:
    tien_chua_vat = 50 * 1984 + 50 * 2050 + 100 * 2380 + (kwh - 200) * 2998
elif kwh <= 400:
    tien_chua_vat = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + (kwh - 300) * 3350
else:
    tien_chua_vat = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + 100 * 3350 + (kwh - 400) * 3460

tong_tien = int(tien_chua_vat * 1.08)
print(f"Ho va ten: {ten}")
print(f"Tien phai tra la: {tong_tien}")
```

#### 9. Chuyển đổi số thành chữ số: Nhập một số nguyên từ 0 đến 9, in ra tên chữ số tương ứng (0 -> "Không", 1 -> "Một", ...) – Xem trước match …case

```python
so = int(input("Nhập một số nguyên (0-9): "))
match so:
    case 0:
        print("Không")
    case 1:
        print("Một")
    case 2:
        print("Hai")
    case 3:
        print("Ba")
    case 4:
        print("Bốn")
    case 5:
        print("Năm")
    case 6:
        print("Sáu")
    case 7:
        print("Bảy")
    case 8:
        print("Tám")
    case 9:
        print("Chín")
    case _:
        print("Số nhập vào không hợp lệ!")
```
