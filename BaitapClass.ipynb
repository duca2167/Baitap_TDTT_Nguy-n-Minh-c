[Cylinder]
Xây dựng lớp Cylinder (hình trụ) gồm: bán kính và chiều cao
(kiểu số thực) với hai phương thức:tính diện tích bề mặt (diện
tích toàn phần) và thể tích của hình trụ
Đầu vào
Đâu vào từ bàn phím gồm một dòng duy nhất chứa bán kính và
chiều cao của hình trụ, cách nhau bởi một dấu cách.
Đầu ra
In ra màn hình diện tích bề mặt và thể tích của hình trụ (sử
dụng lại những hàm vừa viết, π=3.14).
(Kết quả làm tròn đến 2 chữ số sau dấu phẩy).

```python
class Cylinder:
    def __init__(self, radius = 0, height = 0):
        self.radius = radius
        self.height = height
    def surface_area(self):
        return 2 * 3.14 * self.radius * (self.radius + self.height)
    def volume(self):
        return 3.14 * self.radius * self.radius * self.height
    def __str__(self):
        return f"Radius: {self.radius}, Height: {self.height}"
    def __repr__(self):
        return f"{self.radius} - {self.height}"

r, h = map(float, input().split())
c = Cylinder(r, h)
print(c)
c
print(f"{c.surface_area():.2f}")
print(f"{c.volume():.2f}")

```

#### W8A2
[Date]
Xây dựng lớp Date biểu diễn ngày tháng năm với hai phương
thức kiểm tra xem một ngày tháng nào đó có hợp lệ hay không
và phương thức tính ngày kế tiếp của ngày tháng năm nào đó
nếu nó là hợp lệ
Nhập vào một ngày tháng năm, (ngày tháng năm biểu diễn
dưới dạng dd/mm/yyyy). Nếu ngày tháng năm không tồn tại, in
ra dòng chữ INVALID. Nếu là hợp lệ in ra ngày kế tiếp của nó.
Đầu vào
Đầu vào từ bàn phim chứa dữ liệu ngày, tháng, năm dưới dạng
dd/mm/yyyy
.
Đầu ra
In ra màn hình ngày kế tiếp của ngày vừa nhập dưới dạng
dd/mm/yyyy
. Nếu ngày tháng không tồn tại, in ra INVALID

```python
class Date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year
    def checkValid(self):
        max_day = -1
        match self.month:
            case 1|3|5|7|8|10|12:
                max_day = 31
            case 4|6|9|11:
                max_day = 30
            case 2:
                if self.year % 4 == 0 and self.year % 100 != 0 or self.year % 400 == 0:
                    max_day = 29
                else:
                    max_day = 28
        return (1 <= self.day <= max_day and 1 <= self.month <= 12 and 0 <= self.year)
    def __str__(self):
        return f"{self.day:02d}/{self.month:02d}/{self.year:04d}"
    def nextDay(self):
        nextDay = Date(self.day + 1, self.month, self.year)
        if nextDay.checkValid():
            return nextDay
        nextDay.day = 1
        nextDay.month += 1
        if nextDay.checkValid():
            return nextDay
        nextDay.month = 1
        nextDay.year += 1
        return nextDay
day, month, year = map(int, input().split("/"))
date1 = Date(day, month, year)
if date1.checkValid():
    print(date1.nextDay())
else:
    print("INVALID")


    
    
```

Xây dựng lớp ShoppingCart (giỏ hàng) để quản lý các sản

Với các phương thức sau:
1. Khởi tạo với danh sách rỗng
2. Thêm vào giỏ hàng một sản phẩm mới
3. Xóa khỏi giỏ một sản phẩm nào đó (tìm theo tên để xóa)
4. Kiểm tra xem giỏ hàng có rỗng không
5. Tính tổng tiền giỏ hàng
6. Hiển thị toàn bộ sản phẩm trong giỏ

Chương trình dừng khi người dùng nhập "yes" cho câu hỏi "Có
muốn thoát khỏi chương trình không?"

```python
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, name, price):
        product = {"name": name, "price": price}
        self.items.append(product)

    def remove_item(self, name):
        for item in self.items:
            if item["name"] == name:
                self.items.remove(item)
                print(f"Đã xóa {name} ra khỏi khoang chứa")
                return
        print("Không tìm thấy sản phẩm này trong khoang chứa.")

    def is_empty(self):
        return len(self.items) == 0

    def total_price(self):
        total = 0
        for item in self.items:
            total += item["price"]
        return total

    def display_items(self):
        if self.is_empty():
            print("Giỏ hàng đang trống")
        else:
            for item in self.items:
                print(f"{item['name']}: {item['price']} VND")

    def clear_cart(self):
        self.items = []
        print("Đã xóa toàn bộ giỏ hàng")
# Chương trình chính
cart = ShoppingCart()

while True:
    print("\n--- MENU QUẢN LÝ GIỎ HÀNG ---")
    print("2. Thêm sản phẩm")
    print("3. Xóa sản phẩm")
    print("4. Kiểm tra giỏ hàng có rỗng không")
    print("5. Tính tổng tiền")
    print("6. Hiển thị giỏ hàng")
    print("7. Xóa toàn bộ giỏ hàng")
    
    choice = input("Nhập lựa chọn của ông (2-7): ")

    if choice == '2':
        name = input("Tên sản phẩm: ")
        price = float(input("Giá sản phẩm: "))
        cart.add_item(name, price)
    elif choice == '3':
        name = input("Tên sản phẩm muốn xóa: ")
        cart.remove_item(name)
    elif choice == '4':
        if cart.is_empty():
            print("Giỏ hàng đang trống.")
        else:
            print("Giỏ hàng có sản phẩm.")
    elif choice == '5':
        print(f"Tổng tiền: {cart.total_price()} VNĐ")
    elif choice == '6':
        cart.display_items()
    elif choice == '7':
        cart.clear_cart()
    
    thoat = input("Có muốn thoát khỏi chương trình không? (yes/no): ")
    if thoat.lower() == "yes":
        break
```

#### W8A7
Yêu cầu
Lớp Fraction(num, den) (tử, mẫu nguyên, mẫu ≠ 0) với:
Rút gọn phân số trong __init__ (quy ước mẫu > 0).
Các phương thức add/sub/mul/div trả về phân số đã rút gọn.
__str__() in dạng a/b.
Input
a b op c d (tương ứng a/b (op) c/d, op thuộc {+,-,*,/})
Output
Phân số kết quả đã rút gọn

```python
import math
class Fraction:
    def __init__(self, num, den):
        common = math.gcd(num, den)
        self.num = num // common
        self.den = den // common
        if self.den < 0:
            self.num *= -1
            self.den *= -1
    
    def __str__(self):
        return f"{self.num}/{self.den}"
    def __add__(self, other):
        num = self.num * other.den + self.den * other.num
        den = self.den * other.den
        return Fraction(num, den)
    def __sub__(self, other):
        num = self.num * other.den - self.den * other.num
        den = self.den * other.den
        return Fraction(num, den)
    def __mul__(self, other):
        num = self.num * other.num
        den = self.den * other.den
        return Fraction(num, den)
    def __truediv__(self, other):
        num = self.num * other.den
        den = self.den * other.num
        return Fraction(num, den)
a, b, op, c, d = input().split()
f1 = Fraction(int(a), int(b))
f2 = Fraction(int(c), int(d))
if op == "+":
    print(f1 + f2)
elif op == "-":
    print(f1 - f2)
elif op == "*":
    print(f1 * f2)
elif op == "/":
    print(f1 / f2)

    
```

#### W8A9
Yêu cầu
Lớp Student(name) với:
add_score(subject, score)
avg() trả về điểm trung bình.
rank() trả về "Excellent" nếu avg ≥ 8, "Good" nếu 6.5–<8,
"Average" nếu 5–<6.5, ngược lại "Poor".
Input
Dòng 1: name
Dòng 2: n
n dòng tiếp: subject score (score là số thực 0–10)
Output
name avg rank (avg 2 chữ số).

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.dict = {}
    def add_score(self, subject, score):
        if subject not in self.dict:
            self.dict[subject] = score
    def avg(self):
        total = 0
        count = 0
        for score in self.dict.values():
            total += float(score)
            count += 1
        average = total / count
        return average
    def rank(self):
        if self.avg() >= 8:
            return "Excellent"
        elif 6.5 <= self.avg() < 8:
            return "Good"
        elif 5 <= self.avg() < 6.5:
            return "Average"
        else:
            return "Poor"

name = input()
s = Student(name)
n = int(input())
for _ in range(n):
    subject, score = input().split()
    s.add_score(subject, score)
print(f"{name} {s.avg():.2f}", end = " ")
print(s.rank())
            

        
```

#### W8A11
Bạn được cho bốn điểm A, B, C và D trong hệ tọa độ Đề-các 3
chiều.
Nhiệm vụ của bạn là tính và in ra góc giữa hai mặt phẳng tạo
bởi các điểm:
mặt phẳng (A, B, C)
mặt phẳng (B, C, D)
Góc cần in ra được ký hiệu là PHI, và phải dùng độ (degrees),
không dùng radian.
Dạng Input:
Một dòng input chứa 3 số thực, ngăn cách bằng dấu cách, biểu
diễn:
tọa độ X tọa độ Y tọa độ Z của một điểm.
Dạng Output:
In ra góc giữa hai mặt phẳng, chính xác tới hai chữ số thập
phân.

```python
import math
class GeometryCalculator:
    @staticmethod
    def cross_product(u, v):
        return (u[1]*v[2] - u[2]*v[1], u[2]*v[0] - u[0]*v[2], u[0]*v[1] - u[1]*v[0])
    @staticmethod
    def dot_product(u, v):
        return u[0]*v[0] + u[1]*v[1] + u[2]*v[2]
    @staticmethod
    def length(u):
        return math.sqrt(u[0]*u[0] + u[1]*u[1] + u[2]*u[2])
    def calculate_angle(self, A, B, C, D):
        AB = (B[0] - A[0], B[1] - A[1], B[2] - A[2])
        AC = (C[0] - A[0], C[1] - A[1], C[2] - A[2])
        BC = (C[0] - B[0], C[1] - B[1], C[2] - B[2])
        BD = (D[0] - B[0], D[1] - B[1], D[2] - B[2])

        n1 = self.cross_product(AB, AC)
        n2 = self.cross_product(BC, BD)

        cos_phi = abs(self.dot_product(n1, n2)) / (self.length(n1) * self.length(n2))
        return math.degrees(math.acos(cos_phi))
points = []
for _ in range(4):
    point = list(map(float, input().split()))
    points.append(point)
calc = GeometryCalculator()
angle = calc.calculate_angle(points[0], points[1], points[2], points[3])
print(f"{angle:.2f}")
```

[Complex]
Khai báo kiểu dữ liệu biểu diễn số phức \(ComplexNumber\)
bao gồm \(2\) biến \(realPart\) và \(imaginaryPart\) kiểu
\(double\) biểu diễn phần thực và phần ảo của một số phức.
Viết hàm khởi tạo ComplexNumber(double initRealPart, double
initImaginaryPart) để khởi tạo các giá trị phần thực và phần ảo
của một số phức và hàm ComplexNumber
multiply(complexNumber other) trả về tích của số phức hiện tại
và một số phức khác.
Đầu vào
Đầu vào từ bàn phím gồm hai dòng, mỗi dòng gồm hai số thực
cách nhau bởi một dấu cách biểu diễn phần thực và phần ảo
của một số phức. Các số trong đầu vào có giá trị tuyệt đối
không vượt quá \(100\).
Đầu ra
In ra màn hình hai số biểu diễn phần thực và phần ảo của tích
hai số phức.

```python
class ComplexNumber:
    def __init__(self, initRealPart, initImaginaryPart):
        self.realPart = initRealPart
        self.imaginaryPart = initImaginaryPart

    def multiply(self, other):
        r1, i1 = self.realPart, self.imaginaryPart
        r2, i2 = other.realPart, other.imaginaryPart
        
        real = r1 * r2 - i1 * i2
        imag = r1 * i2 + i1 * r2
        
        return ComplexNumber(real, imag)

data1 = input().split()
c1 = ComplexNumber(int(data1[0]), int(data1[1]))

data2 = input().split()
c2 = ComplexNumber(int(data2[0]), int(data2[1]))

result = c1.multiply(c2)
print(f"{result.realPart} {result.imaginaryPart}")
```

Hoàn thiện chương trình dưới đây nhập vào \(3\) cạnh của một
tam giác và in ra diện tích của tam giác đó.
Đầu vào
Một dòng duy nhất từ bàn phím chứa \(3\) số nguyên là độ dài
\(3\) cạnh của tam giác
Đầu ra
In ra màn hình diện tích tam giác (làm tròn đến\(2\) chữ số thập
phân sau dấu phẩy). In ra \(invalid\) nếu \(3\) số nhập vào
không phải là \(3\) cạnh của một tam giác

```python
import math

class Triangle:
    def __init__(self, initEdge1, initEdge2, initEdge3):
        self.edge1 = initEdge1
        self.edge2 = initEdge2
        self.edge3 = initEdge3

    def getArea(self):
        a, b, c = self.edge1, self.edge2, self.edge3
        if a + b > c and a + c > b and b + c > a:
            s = (a + b + c) / 2
            area = math.sqrt(s * (s - a) * (s - b) * (s - c))
            return area
        else:
            return -1

edges = input().split()
t = Triangle(int(edges[0]), int(edges[1]), int(edges[2]))
area = t.getArea()

if area == -1:
    print("invalid")
else:
    print(f"{area:.2f}")
```

#### W8A14 Yêu cầu
Xây dựng lớp Animal với phương thức speak().
Xây dựng hai lớp kế thừa Dog và Cat, mỗi lớp ghi đè speak()
để trả về âm thanh tương ứng:
- Dog -> Woof
- Cat -> Meow
Input
Một dòng chứa tên lớp: Dog hoặc Cat.
Output
In ra kết quả của speak().

```python
class Animal:
    def speak(self):
        pass
class Dog(Animal):
    def speak(self):
        print("Woof")
class Cat(Animal):
    def speak(self):
        print("Meow")
animal = input()
if animal.lower() == "dog":
    dog = Dog()
    dog.speak()
elif animal.lower() == "cat":
    cat = Cat()
    cat.speak()
```

#### W8A15 Yêu cầu
Xây dựng lớp Employee(name, salary).
Xây dựng lớp Manager(name, salary, bonus) kế thừa từ
Employee.
Cả hai lớp đều có phương thức income():
- Employee: trả về salary
- Manager: trả về salary + bonus
Input
Dòng 1: loại đối tượng (Employee hoặc Manager)
Dòng 2: name salary [bonus]
Output
In ra tổng thu nhập của đối tượng.

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    def income(self):
        return self.salary
class Manager(Employee):
    def __init__(self, name, salary, bonus):
        super().__init__(name, salary)
        self.bonus = bonus
    def income(self):
        return self.salary + self.bonus
person = input()
if person.lower() == "manager":
    name, salary, bonus = input().split()
    manager = Manager(name, int(salary), int(bonus))
    print(manager.income())
elif person.lower() == "employee":
    name, salary = input().split()
    employee = Employee(name, int(salary))
    print(employee.income())
```
