#### W7A16
Cho một mảng số nguyên không âm. Gọi
mảng đó là "mảng đẹp" nếu mỗi phần tử
x trong mảng xuất hiện đúng x lần.
Kiểm tra xem mảng có phải là mảng đẹp
hay không. Nếu không hãy xóa đi y số (y
là nhỏ nhất có thể) để mảng còn lại thành
mảng đẹp.

```python
arr = list(map(int, input().split()))
count = {}
for so in arr:
    if so in count:
        count[so] += 1
    else:
        count[so] = 1
tong_so_xoa = 0
for so, so_lan in count.items():
    if so == so_lan:
        tong_so_xoa += 0
    elif so < so_lan:
        tong_so_xoa += so_lan - so
    else:
        tong_so_xoa += so_lan
if tong_so_xoa == 0:
    print("YES")
else:
    print("NO", end = " ")
    print(tong_so_xoa)
        
```

#### W7A17
Tìm phần tử xuất hiện nhiều hơn n/2 lần
trong mảng (nếu có) với độ phức tạp thời
gian là O(n), độ phức tạp bộ nhớ là O(1),
- Boyer–Moore. Nếu không có phần tử
nào thỏa mãn thì không in gì

```python
arr = list(map(int, input().split()))
ung_vien = None
count = 0
for i in range(len(arr)):
    if count == 0:
        ung_vien = arr[i]
        count = 1
    elif arr[i] == ung_vien:
        count += 1
    else:
        count -= 1
for so in arr:
    if so == ung_vien:
        count += 1
if count > len(arr) / 2:
    print(ung_vien)
else:
    print(" ")
```

#### W7A19
Merge 2 mảng đã sắp xếp (có thể tăng
dần, có thể giảm dần) mà không dùng
bất cứ thuật toán sắp xếp nào thành một
mảng cũng có cùng thứ tự sắp xếp như 2
mảng ban đầu

```python
lst1 = list(map(int, input().split()))
lst2 = list(map(int, input().split()))

tang = True
for i in range(len(lst1) - 1):
    if lst1[i] > lst1[i+1]:
        tang = False
        break

i = 0
j = 0
result = []

if tang:
    while i < len(lst1) and j < len(lst2):
        if lst1[i] <= lst2[j]:
            result.append(lst1[i])
            i += 1
        else:
            result.append(lst2[j])
            j += 1
    while i < len(lst1):
        result.append(lst1[i])
        i += 1
    while j < len(lst2):
        result.append(lst2[j])
        j += 1
else:
    while i < len(lst1) and j < len(lst2):
        if lst1[i] >= lst2[j]:
            result.append(lst1[i])
            i += 1
        else:
            result.append(lst2[j])
            j += 1
    while i < len(lst1):
        result.append(lst1[i])
        i += 1
    while j < len(lst2):
        result.append(lst2[j])
        j += 1

print(*result)
```

#### W7A20
Cho dãy N số nguyên dương, tính tổng
các số có k chữ số. Dòng đầu ghi N số
nguyên dương, dòng sau ghi số nguyên
dương k 

```python
def dem_chu_so(n):
    count = 0
    while n >= 1:
        count += 1
        n //= 10
    return count
arr = list(map(int, input().split()))
k = int(input())
tong = 0
for so in arr:
    if dem_chu_so(so) == k:
        tong += so
print(tong)
```

#### W7A22
Cho N tên sinh viên gồm cả họ và tên,
tìm xem tên nào hay gặp nhất. Nếu có
nhiều tên cùng số lần gặp thì lấy tên có
thứ tự từ điển nhỏ nhất

```python
n = int(input())
names = []
for _ in range(n):
    name = input().split()
    names.append(name)

count = {}
for i in range(n):
    ten = names[i][-1]
    if ten in count:
        count[ten] += 1
    else:
        count[ten] = 1

max_lan = 0
ten_nhieu_nhat = ""

for ten, so_lan in count.items():
    if so_lan > max_lan:
        max_lan = so_lan
        ten_nhieu_nhat = ten
    elif so_lan == max_lan:
        if ten.lower() < ten_nhieu_nhat.lower():
            ten_nhieu_nhat = ten

print(ten_nhieu_nhat)
```

#### W7A24
Cho xâu kí tự, xóa các từ giống nhau liền
kề, chỉ giữ lại từ đầu tiên

```python
s = input()

words = []
current_word = ""

for char in s:
    if char != " ":
        current_word += char
    else:
        if current_word != "":
            words.append(current_word)
            current_word = ""
if current_word != "":
    words.append(current_word)

if not words:
    print("")
else:
    result = [words[0]]
    for i in range(1, len(words)):
        if words[i] != words[i - 1]:
            result.append(words[i])
            
    output = ""
    for i in range(len(result)):
        if i == len(result) - 1:
            output += result[i]
        else:
            output += result[i] + " "
            
    print(output)
```

#### W7A25
Tìm số có đúng k chữ số lớn nhất có thể
trong một số có n chữ số (coi đầu vào là
một số rất lớn - str). Không được hoán
đổi vị trí các chữ số

```python
s = input()
k = int(input())

result = ""
start_idx = 0

for i in range(k):
    max_char = " "
    max_idx = -1
    
    end_idx = len(s) - (k - i)
    
    for j in range(start_idx, end_idx + 1):
        if s[j] > max_char:
            max_char = s[j]
            max_idx = j
            
    result += max_char
    start_idx = max_idx + 1

print(result)
```

#### W7A27
Cho một dãy ngoặc ()(){}{}[][]..., kiểm tra
xem dãy ngoặc đã cho có hợp lệ hay
không

```python
def kiem_tra_ngoac(xau):
    kho = []
    cap_ngoac = {')': '(', ']': '[', '}': '{'}
    
    for ky_tu in xau:
        if ky_tu in '([{':
            kho.append(ky_tu)
        elif ky_tu in ')]}':
            if not kho:
                return False
            gan_nhat = kho.pop()
            if gan_nhat != cap_ngoac[ky_tu]:
                return False
                
    if kho:
        return False
    return len(kho) == 0
day_ngoac = input("Nhập một chuỗi ngoặc: ")
if kiem_tra_ngoac(day_ngoac):
    print("YES")
else:
    print("NO")
```

#### W7A28
Bà̀i toán Tìm dãy các hợp số dài nhất liên
tiếp không chứa số nguyên nào <= N (N
là số nguyên dương cho trước). Nếu có
nhiều dãy cùng độ dài thì lấy dãy nhỏ
nhất (gặp đầu tiên từ trái qua)

```python
n = int(input())

is_prime = [True] * (n + 1)
if n >= 0: is_prime[0] = False
if n >= 1: is_prime[1] = False

for i in range(2, n + 1):
    if is_prime[i]:
        for j in range(i * i, n + 1, i):
            is_prime[j] = False

max_len = 0
best_start = -1

current_len = 0
current_start = -1

for i in range(2, n + 1):
    if not is_prime[i]:
        if current_len == 0:
            current_start = i
        current_len += 1
    else:
        if current_len > max_len:
            max_len = current_len
            best_start = current_start
        current_len = 0
        current_start = -1

if current_len > max_len:
    max_len = current_len
    best_start = current_start

if max_len == 0:
    print("")
else:
    result = []
    for i in range(best_start, best_start + max_len):
        result.append(i)
    print(*result, sep=", ")
```

#### W7A29
Tìm Dãy con tăng dần liên tục dài nhất
của một danh sách nhập vào từ bàn
phím. Lấy dãy đầu tiên nếu có nhiều dãy
có số phần tử bằng nhau

```python
mang_so = list(map(int, input().split()))

if not mang_so:
    print("")
else:
    day_lon_nhat = [mang_so[0]]
    day_hien_tai = [mang_so[0]]
    
    for i in range(1, len(mang_so)):
        if mang_so[i] > mang_so[i - 1]:
            day_hien_tai.append(mang_so[i])
        else:
            if len(day_hien_tai) > len(day_lon_nhat):
                day_lon_nhat = day_hien_tai
            day_hien_tai = [mang_so[i]]
            
    if len(day_hien_tai) > len(day_lon_nhat):
        day_lon_nhat = day_hien_tai
        
    print(*day_lon_nhat)
```

#### W7A30
Tìm Dãy con tăng dần KHÔNG liên tục
dài nhất của một danh sách nhập vào từ
bàn phím

```python
mang_so = list(map(int, input().split()))

if not mang_so:
    print("")
else:
    so_phan_tu = len(mang_so)
    
    cac_day_con = []
    for i in range(so_phan_tu):
        cac_day_con.append([mang_so[i]])
        
    for i in range(1, so_phan_tu):
        for j in range(i):
            if mang_so[j] < mang_so[i]:
                if len(cac_day_con[j]) + 1 > len(cac_day_con[i]):
                    cac_day_con[i] = cac_day_con[j] + [mang_so[i]]
                    
    day_dai_nhat = cac_day_con[0]
    for i in range(1, so_phan_tu):
        if len(cac_day_con[i]) > len(day_dai_nhat):
            day_dai_nhat = cac_day_con[i]
            
    print(*day_dai_nhat)
```

#### W7A31
Đổi chỗ đúng 2 chữ số để được số nhỏ
nhất có thể

```python
s = input()

mang_ky_tu = []
for char in s:
    mang_ky_tu.append(char)

so_luong = len(mang_ky_tu)
tim_thay = False

for i in range(so_luong):
    vi_tri_nho_nhat = i
    for j in range(so_luong - 1, i, -1):
        if mang_ky_tu[j] < mang_ky_tu[vi_tri_nho_nhat]:
            if i == 0 and mang_ky_tu[j] == "0":
                continue
            vi_tri_nho_nhat = j
            
    if vi_tri_nho_nhat != i:
        temp = mang_ky_tu[i]
        mang_ky_tu[i] = mang_ky_tu[vi_tri_nho_nhat]
        mang_ky_tu[vi_tri_nho_nhat] = temp
        tim_thay = True
        break

if not tim_thay:
    for i in range(so_luong - 1, 0, -1):
        if mang_ky_tu[i] < mang_ky_tu[i - 1]:
            temp = mang_ky_tu[i]
            mang_ky_tu[i] = mang_ky_tu[i - 1]
            mang_ky_tu[i - 1] = temp
            break

ket_qua = ""
for char in mang_ky_tu:
    ket_qua += char

print(ket_qua)
```
