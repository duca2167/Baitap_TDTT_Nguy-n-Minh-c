#### W3A3 Kiểm tra xem một số có phải là lũy thừa của 2 hay không
## Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2 hay không.
#### Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0

```python
n = int(input("Nhập một số nguyên dương: "))
if n > 0 and (n & (n - 1) == 0):
    print(f"{n} là lũy thừa của 2")
else:
    print(f"{n} không phải là lũy thừa của 2")
```

```python
a=int(input("Nhập số nguyên dương a:"))
b=int(input("Nhập số nguyên dương b:"))
c=int(input("Nhập số nguyên dương c:"))
if a+b>c and a+c>b and b+c>a:
    if a==b==c:
        print("Tam giác đều")
    elif a==b or a==c or b==c:
        print("Tam giác cân")
    else:
        print("Tam giác thường")
else:
    print("Không phải là tam giác")
```

#### - 0-50 kWh: 1500đ/kWh
#### - 51-100 kWh: 2000đ/kWh
#### - 100 kWh: 3000đ/kWh

```python
x=int(input("Nhập số tiền điện kWh: "))
if 0<=x<=50:
    print(f"Số tiền điện phải trả là: {x*1500} VND")
elif 51<=x<=100:
    print(f"Số tiền điện phải trả là: {50*1500 + (x-50)*2000} VND")
else:
    print(f"Số tiền điện phải trả là: {50*1500 + 50*2000 + (x-100)*3000} VND")
```
