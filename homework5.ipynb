Viết hàm kiểm tra một số có phải số Amstrong hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
#Hàm xét số armstrong
def armstrong(n):
    tong = 0
    so_mu = len(str(n))
    for i in str(n):
        tong += int(i)**so_mu
    if tong == n:
        return "True"
    else:
        return "False"
    
# Chương trình chính
x = int(input())
print(armstrong(x))
```

```python
def detect_number_type(n):
    try:
        num_float = float(n)
        # Xét xem có phải số không 
        if num_float == int(num_float):
            return "int"
        else:
            return "float"
    except ValueError:
        return "neither"

# Chương trình chính
s = input()
print(detect_number_type(s))
```

#### W5A23: Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0.

```python
import math

def quadratic_roots_count(a, b, c):
    if a == 0:
        if b == 0:
            if c == 0:
                return "Phương trình vô số nghiệm"
            else:
                return "Phương trình vô nghiệm"
        x = -c / b
        return f"x={x}", "Số nghiệm thực là 1"

    delta = b * b - 4 * a * c
    if delta < 0:
        return "Phương trình không có nghiệm thực nào"
    if delta == 0:
        x = round(-b / (2 * a), 3)
        return x, "Số nghiệm thực là 1"

    x1 = round((-b + math.sqrt(delta)) / (2 * a), 3)
    x2 = round((-b - math.sqrt(delta)) / (2 * a), 3)
    return x1, x2, "Số nghiệm thực là 2"

a = float(input())
b = float(input())
c = float(input())
print(quadratic_roots_count(a, b, c))
```

#### W5A24: Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.

```python
def are_collinear(x1, y1, x2, y2, x3, y3):
    # vectơ u = (x2 - x1, y2 - y1)
    # vectơ v = (x3 - x1, y3 - y1)
    # kiểm tra định thức u x v = 0 để xác định ba điểm có thẳng hàng hay không
    eps = 1e-9
    dx1 = x2 - x1
    dy1 = y2 - y1
    dx2 = x3 - x1
    dy2 = y3 - y1

    if abs(dx1) < eps and abs(dy1) < eps and abs(dx2) < eps and abs(dy2) < eps:
        return "Ba điểm trùng nhau"
    if abs(dx1 * dy2 - dy1 * dx2) < eps:
        return "Ba điểm thẳng hàng"
    return "Ba điểm không thẳng hàng"


def read_point(prompt):
    while True:
        try:
            x, y = map(float, input(prompt).split())
            return x, y
        except ValueError:
            print("Nhập sai định dạng. Vui lòng nhập hai số thực cách nhau bằng khoảng trắng.")

x1, y1 = read_point("Nhập tọa độ điểm 1: ")
x2, y2 = read_point("Nhập tọa độ điểm thứ 2: ")
x3, y3 = read_point("Nhập tọa độ điểm thứ 3: ")
print(are_collinear(x1, y1, x2, y2, x3, y3))
```

#### W5A25
Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay
không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3),
(0,3)], p = (2,1)

```python
def cross_product(a, b, c):
    """
    Tính tích có hướng của hai vectơ AB và AC.
    Công thức: (Bx-Ax)*(Cy-Ay) - (By-Ay)*(Cx-Ax)
    """
    return (b[0] - a[0]) * (c[1] - a[1]) - (b[1] - a[1]) * (c[0] - a[0])

def point_in_quadrilateral(quad, p):
    # Duyệt qua 4 cạnh của tứ giác: (quad[0], quad[1]), (quad[1], quad[2]), ...
    n = len(quad)
    dau_duong = False
    dau_am = False
    
    for i in range(n):
        # Lấy hai đỉnh liên tiếp để tạo thành một cạnh
        p1 = quad[i]
        p2 = quad[(i + 1) % n] # Dùng % n để quay lại đỉnh đầu tiên khi ở đỉnh cuối
        
        # Tính tích có hướng của cạnh (p1, p2) và điểm p
        cp = cross_product(p1, p2, p)
        
        if cp > 0:
            dau_duong = True
        elif cp < 0:
            dau_am = True
            
        # Nếu tích có hướng có cả dương và âm, điểm nằm ngoài
        if dau_duong and dau_am:
            return False
            
    # Nếu chỉ toàn dương hoặc toàn âm (hoặc có bằng 0 - nằm trên biên)
    return True

# Chạy thử ví dụ của Đức
quad_test = [(0,0), (4,0), (4,3), (0,3)]
p_test = (2,1)
print(f"Kết quả kiểm tra: {point_in_quadrilateral(quad_test, p_test)}") # Trả về True
```

#### W5A28: Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).

```python
import math

def tinh_goc_abc(a, b, c):
    goc_ba = math.atan2(a[1] - b[1], a[0] - b[0])
    goc_bc = math.atan2(c[1] - b[1], c[0] - b[0])
    
    goc_do = abs(math.degrees(goc_bc - goc_ba))
    
    if goc_do > 180:
        goc_do = 360 - goc_do
        
    return goc_do


print(tinh_goc_abc((0, 3), (0, 0), (4, 0)))
```

```python
import math

def tinh_goc_abc(a, b, c):
    # Tính tọa độ vectơ BA và BC
    ba_x, ba_y = a[0] - b[0], a[1] - b[1]
    bc_x, bc_y = c[0] - b[0], c[1] - b[1]
    
    # Tính tích vô hướng và độ dài
    tich_vo_huong = ba_x * bc_x + ba_y * bc_y
    do_dai_ba = math.sqrt(ba_x**2 + ba_y**2)
    do_dai_bc = math.sqrt(bc_x**2 + bc_y**2)
    
    # Tính cosin và chuyển sang độ
    cos_abc = tich_vo_huong / (do_dai_ba * do_dai_bc)
    
    # Dùng hàm acos để tìm góc (radian) rồi đổi sang độ
    goc_radian = math.acos(cos_abc)
    return math.degrees(goc_radian)


print("Nhập tọa độ các điểm (ví dụ: 0 3):")
A = tuple(map(float, input("Nhập tọa độ điểm A: ").split()))
B = tuple(map(float, input("Nhập tọa độ điểm B (đỉnh góc): ").split()))
C = tuple(map(float, input("Nhập tọa độ điểm C: ").split()))


ket_qua = tinh_goc_abc(A, B, C)
print(f"Số đo góc ABC là: {ket_qua:.2f} độ")
```

#### W5A30: Viết hàm tính hệ số nhị thức C(n, k).

```python
def tinh_to_hop(n, k):
    if k < 0 or k > n:
        return 0
    if k == 0 or k == n:
        return 1
    
    # Tối ưu: C(n, k) = C(n, n-k). Ta chọn số nhỏ hơn để tính cho nhanh.
    if k > n // 2:
        k = n - k
        
    tu_so = 1
    mau_so = 1
    for i in range(k):
        tu_so *= (n - i)
        mau_so *= (i + 1)
        
    return tu_so // mau_so


n_input = int(input("Nhập n: "))
k_input = int(input("Nhập k: "))

print(f"Kết quả C({n_input}, {k_input}) là: {tinh_to_hop(n_input, k_input)}")
```

#### W5A33
Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa,
giá km, phí chờ, phụ phí ban đêm và giảm giá

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    # 1. Xác định giá mỗi km dựa trên loại xe
    if car_type == 4:
        gia_per_km = 12000
    else:
        gia_per_km = 15000
        
    # 2. Tính tiền theo khoảng cách và thời gian chờ
    tien_xe = km * gia_per_km
    tien_cho = wait_minutes * 500
    tong_chua_phu_phi = tien_xe + tien_cho
    
    # 3. Tính phụ phí ban đêm (ví dụ từ 22h đêm đến 5h sáng)
    if start_hour >= 22 or start_hour < 5:
        tong_chua_phu_phi *= 1.1  # Tăng 10%
        
    return tong_chua_phu_phi 
print("--- HỆ THỐNG TÍNH CƯỚC TAXI THÔNG MINH ---")
loai_xe = int(input("Nhập loại xe (4 hoặc 7 chỗ): "))
gio_di = int(input("Nhập giờ bắt đầu đi (0-23): "))
so_km = float(input("Nhập số km đã đi: "))
phut_cho = int(input("Nhập số phút chờ: "))


tong_bill = smart_taxi_fare(loai_xe, gio_di, so_km, phut_cho)

print(f"\nTổng tiền cước của bạn là: {tong_bill:,.0f} VNĐ")
```

#### W5A34
Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau
Nếu n ≤ 0 trả về 0.

```python
def sum_1_to_n(n):
    tong = 0
    for i in range(1, n+1):
        i_str = str(i)
        if len(i_str) == len(set(i_str)):
            tong += i
    return tong 
n=int(input())
print(sum_1_to_n(n))
```
