#### W6A1
Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng giữ
nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.
Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3]

```python
lst_ban_dau = list(map(int, input("Nhập vào một dãy số nguyên, cách nhau bởi dấu cách: ").split()))
cac_so_da_xuat_hien = {}
lst_moi = []
for so in lst_ban_dau:
    if so not in cac_so_da_xuat_hien:
        cac_so_da_xuat_hien[so] = True
        lst_moi.append(so)
print(lst_moi)
```

sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp

#### W6A4
Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo
dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key lặp
lại, thêm vào list của key đó

```python
chuoi = input("Nhập các cặp key:value, cách nhau bởi dấu cách: ")
lst = chuoi.split()
tu_dien = {}
for item in lst:
    key, value = item.split(":")
    if key in tu_dien:
        tu_dien[key].append(value)
    else:
        tu_dien[key] = [value]
print(tu_dien)
```

#### W6A5
Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 'zeros':
...} (đếm số >0, <0, =0)

```python
dong_so_nguyen = list(map(int, input().split()))
dct ={
    'positives' : 0,
    'negatives' : 0,
    'zeros' : 0
}

for so in dong_so_nguyen:
    if so > 0:
        dct['positives'] += 1
    elif so < 0:
        dct['negatives'] += 1
    else:
        dct['zeros'] += 1

print(dct)
```

#### W6A9
Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một dictionary mới
trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ được cộng lại, nếu chỉ có
trong một dict thì giữ nguyên. In ra dictionary mới theo thứ tự từ điển tăng dần của
key.

```python
def nhap_dict(prompt):
    print(prompt)
    data = input("Nhập các cặp key:value: ").split()
    d = {}
    for item in data:
        k, v = item.split(':')
        d[k] = int(v) # Chuyển value sang kiểu số để cộng được
    return d

dict1 = nhap_dict("--- Nhập Dictionary 1 ---")
dict2 = nhap_dict("--- Nhập Dictionary 2 ---")

dict_tong = dict(dict1) 

for key, value in dict2.items():
    if key in dict_tong:
        dict_tong[key] += value 
    else:
        dict_tong[key] = value  
dict_sap_xep = {}
keys = list(dict_tong.keys())
n = len(keys)
for i in range(n-1):
    for j in range(i+1, n):
        if keys[i] > keys[j]:
            keys[i], keys[j] = keys[j], keys[i]
for key in keys:
    dict_sap_xep[key] = dict_tong[key]

print(dict_sap_xep)
```

```python
def tao_dict(lst):
    dct = {}
    for chuoi in lst:
        k, v = chuoi.split(":")
        if k in dct:
            dct[k].append(int(v))
        else:
            dct[k] = [int(v)]
    return dct

def merge_dicts(dict1, dict2):
    dict_moi = {}
    for key in set(dict1.keys()).union(set(dict2.keys())):
        if key in dict1 and key in dict2:
            dict_moi[key] = dict1[key] + dict2[key]
        elif key in dict1:
            dict_moi[key] = dict1[key]
        else:
            dict_moi[key] = dict2[key]
    return dict_moi

dict1 = tao_dict(input("Nhập các cặp key:value, cách nhau bởi dấu cách: ").split())
dict2 = tao_dict(input("Nhập các cặp key:value, cách nhau bởi dấu cách: ").split())
dict_moi = merge_dicts(dict1, dict2)

sorted_dict = dict(sorted(dict_moi.items(), key=lambda item: item[0]))
print(sorted_dict)
```

#### W6A10
Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên dương k. In ra
list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là một tuple, hai số trong tuple
có thể giống nhau, thứ tự xuất hiện trong danh sách của số thứ nhất nhỏ hơn thứ tự của
số thứ hai. Các tuple trong list được sắp xếp theo thứ tự tăng dần của phần tử đầu tiên
trong tuple.

```python
dong_so_nguyen = list(map(int, input().split()))
k = int(input("Nhập số nguyên k: "))
lst = []
for i in range(len(dong_so_nguyen)-1):
    for j in range(i+1, len(dong_so_nguyen)):
        if dong_so_nguyen[i] + dong_so_nguyen[j] == k:
            if (dong_so_nguyen[i], dong_so_nguyen[j]) not in lst:
                lst.append((dong_so_nguyen[i], dong_so_nguyen[j]))
            else:
                continue
print(sorted(lst))
```
