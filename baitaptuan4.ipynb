#### W4A1: [SumOfNumBers] Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để tính tống các số nguyên từ 1 đến n. (n <= 1000)

```python
def tinhtong(n):
    tong = 0
    for i in range(1, n + 1):
        tong += i
    return tong
n = int(input("Nhập n (0-1000): "))
print("Tổng các số từ 1 đến", n, "là:", tinhtong(n))
```

#### W4A7: [LargestPrimeFactor] Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n.

```python
n = int(input("Nhập vào số nguyên dương n (n>=2): "))

uoc_so_nguyen_to_max = 0

# Duyệt tìm các ước số của n
for i in range(2, n + 1):
    if n % i == 0:
        # Kiểm tra xem ước số i này có phải số nguyên tố không
        la_so_nguyen_to = True
        for z in range(2, int(i**0.5) + 1): # Chỉ cần kiểm tra đến căn bậc 2 của i
            if i % z == 0:
                la_so_nguyen_to = False
                break 
        
        # Nếu là số nguyên tố, cập nhật số lớn nhất
        if la_so_nguyen_to:
            uoc_so_nguyen_to_max = i

print("Ước số nguyên tố lớn nhất của số đã nhập là:", uoc_so_nguyen_to_max) 
```

#### W4A8: ReverseAddPalindrome] Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số đối xứng (palindrome). In ra số bước và giá trị palindrome thu được

```python
n = int(input("Nhập số nguyên dương n: "))

# Tạo một hàm để đảo ngược số
def reverse(n):
    so_nguoc = 0
    temp = n
    while temp > 0:
        chu_so_cuoi = temp % 10
        so_nguoc = so_nguoc * 10 + chu_so_cuoi
        temp = temp // 10
    return so_nguoc

so_buoc = 0
# Một số là đối xứng khi n bằng chính số đảo ngược của nó
while n != reverse(n):
    n +=reverse(n)
    so_buoc += 1

print(so_buoc, n)
```

#### W4A10: [LongestCollatzUpToN] Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số nhỏ nhất.

```python
n = int(input("Nhập n: "))
so_tot_nhat = 0
do_dai_max = 0
for i in range(1, n + 1):
    x = i # Giữ nguyên i để không làm hỏng vòng lặp, dùng x để tính toán
    do_dai_hien_tai = 1 # Bắt đầu tính từ chính số đó
    
    # Dùng while để xét về Collatz
    while x != 1:
        if x % 2 == 0:
            x //= 2 
        else:
            x = 3 * x + 1 
        do_dai_hien_tai += 1 
    
    # Kiểm tra xem số này có phải độ dài Collatz lớn nhất không
    if do_dai_hien_tai > do_dai_max:
        do_dai_max = do_dai_hien_tai
        so_tot_nhat = i

print(so_tot_nhat, do_dai_max)
```

#### 

```python
a=int(input("Nhập số nguyên dương đầu tiên: "))
b=int(input("Nhập số nguyên dương thứ hai: "))

#Hàm liệt kê các ước số của hai số  và xét xem có phải thân thiết hay không
def uocso(a, b):
    uocsoa=[]
    uocsob=[]
    for i in range(1, a):
        if a % i == 0:
            uocsoa.append(i)
    for i in range(1, b):
        if b % i == 0:
            uocsob.append(i)
            
    if sum(uocsoa) == b and sum(uocsob) == a:
        return "true"
    else:
        return "false"
print(uocso(a, b))




```

#### W4A14: [GCD] Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình tìm ước chung lớn nhất của hai số m và n

```python
m=int(input("Nhập số nguyên dương m: "))
n=int(input("Nhập số nguyên dương n: "))
uocsom=[]
uocson=[]
uocsochung=[]

#Tìm ước số của m và n
for i in range(1, m+1):
    if m % i == 0:
        uocsom.append(i)
for i in range(1, n+1):
    if n % i == 0:
        uocson.append(i)

#Tìm ước số chung
for i in uocsom:
    for z in uocson:
        if z == i:
            uocsochung.append(i)

#Tìm ước chung lớn nhất
p=max(uocsochung)
print(f"Ước chung lớn nhất của {m} và {n} là: {p}")
```
