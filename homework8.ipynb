#### W8A16 Yêu cầu
Xây dựng lớp Shape với phương thức area().
Xây dựng lớp Rectangle(w, h) và Circle(r) kế thừa từ Shape, ghi đè area().
Lấy π = 3.14.
Input
Dòng 1: loại hình (RECTANGLE hoặc CIRCLE)
Dòng 2: tham số tương ứng.
Output
In ra diện tích với 2 chữ số thập phân.

```python
class Shape:
    def area(self):
        return 0

class Rectangle(Shape):
    def __init__(self, w, h):
        self.w = w
        self.h = h
        
    def area(self):
        return self.w * self.h

class Circle(Shape):
    def __init__(self, r):
        self.r = r
        
    def area(self):
        return 3.14 * (self.r ** 2)

loai_hinh = input()

if loai_hinh == "RECTANGLE":
    w, h = map(float, input().split())
    hinh = Rectangle(w, h)
elif loai_hinh == "CIRCLE":
    r = float(input())
    hinh = Circle(r)

dien_tich = hinh.area()
print(f"{dien_tich:.2f}")
```

#### W8A17 Yêu cầu
Xây dựng lớp Vehicle(brand, speed).
Xây dựng lớp Car(brand, speed, seats) kế thừa từ Vehicle.
Viết phương thức info() cho mỗi lớp:
- Vehicle: brand-speed
- Car: brand-speed-seats
Input
Dòng 1: loại đối tượng (Vehicle hoặc Car)
Dòng 2: dữ liệu tương ứng.
Output
In ra chuỗi info().

```python
class Vehicle:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed
        
    def info(self):
        return f"{self.brand}-{self.speed}"

class Car(Vehicle):
    def __init__(self, brand, speed, seats):
        Vehicle.__init__(self, brand, speed)
        self.seats = seats
        
    def info(self):
        return f"{self.brand}-{self.speed}-{self.seats}"

loai_doi_tuong = input()

if loai_doi_tuong == "Vehicle":
    du_lieu = input().split()
    hang_xe = du_lieu[0]
    toc_do = du_lieu[1]
    thuc_the = Vehicle(hang_xe, toc_do)
    
elif loai_doi_tuong == "Car":
    du_lieu = input().split()
    hang_xe = du_lieu[0]
    toc_do = du_lieu[1]
    so_ghe = du_lieu[2]
    thuc_the = Car(hang_xe, toc_do, so_ghe)

print(thuc_the.info())
```

####  W8A18 Yêu cầu
Xây dựng lớp Polygon với phương thức perimeter().
Xây dựng lớp Triangle(a, b, c) kế thừa từ Polygon và ghi đè perimeter() trả về a+b+c.
Input
Một dòng chứa 3 số nguyên a b c.
Output
In ra chu vi tam giác.

```python
class Polygon:
    def perimeter(self):
        return 0

class Triangle(Polygon):
    def __init__(self, canh_a, canh_b, canh_c):
        self.a = canh_a
        self.b = canh_b
        self.c = canh_c
        
    def perimeter(self):
        return self.a + self.b + self.c

a, b, c = map(int, input().split())
tam_giac = Triangle(a, b, c)
print(tam_giac.perimeter())
```

Dòng 1: Person hoặc Student

In ra kết quả describe().

#### W8A20 Yêu cầu
Xây dựng lớp Account(balance) với hai phương thức deposit(x) và get_balance().
Xây dựng lớp SavingAccount(balance, rate) kế thừa từ Account, thêm phương thức
add_interest() để cộng balance*rate vào số dư.
Input
balance rate
Output
Số dư sau khi cộng lãi, in 2 chữ số thập phân.

```python
class Account:
    def __init__(self, balance):
        self.balance = balance
        
    def deposit(self, x):
        self.balance += x
        
    def get_balance(self):
        return self.balance

class SavingAccount(Account):
    def __init__(self, balance, rate):
        Account.__init__(self, balance)
        self.rate = rate
        
    def add_interest(self):
        self.balance += self.balance * self.rate

balance, rate = map(float, input().split())
savings_account = SavingAccount(balance, rate)
savings_account.add_interest()
print(f"{savings_account.get_balance():.2f}") 
```

#### W8A21 Yêu cầu
Xây dựng lớp Printer với phương thức print_value(x).
Xây dựng lớp BinaryPrinter kế thừa từ Printer, ghi đè print_value(x) để in biểu diễn nhị phân của
số nguyên dương x.
Input
Một số nguyên dương n.
Output
In ra dạng nhị phân của n.

```python
class Printer:
    def print_value(self, x):
        print(x)

class BinaryPrinter(Printer):
    def print_value(self, x):
        if x == 0:
            print("0")
            return
            
        chuoi_nhi_phan = ""
        while x > 0:
            so_du = x % 2
            chuoi_nhi_phan = str(so_du) + chuoi_nhi_phan
            x = x // 2
            
        print(chuoi_nhi_phan)

n = int(input())
may_in = BinaryPrinter()
may_in.print_value(n) 
```

#### W8A22 Yêu cầu
Xây dựng lớp Matrix2x2(a, b, c, d) biểu diễn ma trận:
[a b]
[c d]
Viết phương thức trace() trả về a + d và determinant() trả về ad - bc.
Input
Một dòng chứa 4 số nguyên a b c d.
Output
In ra trace và determinant trên cùng một dòng.

```python
class Matrix2x2:
    def __init__(self, a, b, c, d):
        self.a = a
        self.b = b
        self.c = c
        self.d = d
    def trace(self):
        return self.a + self.d
    def determinant(self):
        return self.a * self.d - self.b * self.c
a, b, c, d = map(int, input().split())
ma_tran = Matrix2x2(a, b, c, d)
print(ma_tran.trace(), end = " ")
print(ma_tran.determinant())
```

#### W8A25 Yêu cầu
Xây dựng lớp PolynomialSum cho hai đa thức bậc nhất:
P(x) = ax + b, Q(x) = cx + d.
Viết phương thức sum_str() trả về chuỗi biểu diễn đa thức tổng dưới dạng kx + m.
Input
a b c d
Output
Chuỗi của đa thức P(x)+Q(x).

```python
class PolynomialSum:
    def __init__(self, a, b, c, d):
        self.a = a
        self.b = b
        self.c = c
        self.d = d
        
    def sum_str(self):
        k = self.a + self.c
        m = self.b + self.d
        return f"{k}x + {m}"

a, b, c, d = map(int, input().split())
da_thuc = PolynomialSum(a, b, c, d)
print(da_thuc.sum_str())
```

#### W8A27 Yêu cầu
Xây dựng lớp Product(name, price).
Xây dựng lớp DiscountProduct(name, price, percent) kế thừa từ Product, ghi đè phương thức
final_price() để trả về giá sau giảm.
Công thức: price * (100 - percent) / 100
Input
name price percent
Output
In ra giá cuối cùng với 2 chữ số thập phân.

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    def final_price(self):
        return self.price

class DiscountProduct(Product):
    def __init__(self, name, price, percent):
        Product.__init__(self, name, price)
        self.percent = percent
    def final_price(self):
        final = self.price * (100 - self.percent) / 100
        return final

name, price, percent = input().split()
price = float(price)
percent = float(percent)
san_pham = DiscountProduct(name, price, percent)
print(f"{san_pham.final_price():.2f}")
```

#### W8A28 Yêu cầu
Xây dựng lớp Counter có thuộc tính count ban đầu bằng 0 và các phương thức inc(), dec(),
value().
Sau đó thực hiện dãy thao tác trên đối tượng.
Input
Dòng 1: n
Dòng 2: n thao tác, mỗi thao tác là INC hoặc DEC, cách nhau bởi dấu cách.
Output
Giá trị cuối cùng của count.

```python
class Counter:
    def __init__(self):
        self.count = 0
    def inc(self):
        self.count += 1
    def dec(self):
        self.count -= 1
    def value(self):
        return self.count

n = int(input())
dem = Counter()
danh_sach_thao_tac = input().split()
for operation in danh_sach_thao_tac:
    if operation == "INC":
        dem.inc()
    elif operation == "DEC":
        dem.dec()
print(dem.value())
```

```python
class Time:
    def __init__(self, h, m, s):
        self.h = h
        self.m = m
        self.s = s
    def to_seconds(self):
        return self.h * 3600 + self.m * 60 + self.s
h, m, s = map(int, input().split())
thoi_gian = Time(h, m, s)
print(thoi_gian.to_seconds())
```

#### W8A30 Yêu cầu
Xây dựng lớp Stack sử dụng danh sách để cài đặt ngăn xếp với các phương thức push(x),
pop(), top().
Dữ liệu đảm bảo mọi lệnh pop và top đều hợp lệ.
Input
Dòng 1: n
n dòng tiếp theo là một trong ba lệnh:
PUSH x
POP
TOP
Output
Mỗi lệnh TOP in ra giá trị ở đỉnh ngăn xếp trên một dòng.

```python
class Stack:
    def __init__(self):
        self.elements = []
    def push(self, x):
        self.elements.append(x)
    def pop(self):
        if len(self.elements) == 0:
            return None
        return self.elements.pop()
    def top(self):
        if len(self.elements) == 0:
            return None
        return self.elements[-1]

n = int(input())
stack = Stack()
for _ in range(n):
    dong_lenh = input().split()
    operation = dong_lenh[0]
    if operation == "PUSH":
        x = int(dong_lenh[1])
        stack.push(x)
    elif operation == "POP":
        stack.pop()
    elif operation == "TOP":
        print(stack.top())
```
