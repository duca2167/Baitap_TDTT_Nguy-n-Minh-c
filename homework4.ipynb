#### W4A19 [EvenNumBers] Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để in ra các số dương chẵn nhỏ hơn hoặc bằng n (n <= 1000)

```python
# DÙNG VÒNG LẶP FOR
n=int(input("Nhập số nguyên dương n ( n<=1000): "))
for i in range(2, n):
    if i % 2 == 0:
        print(i, end=" ")

```

```python
# DÙNG VÒNG LẶP WHILE
n=int(input("Nhập số nguyên dương n (n<=1000): "))
s=0
while s <= n and n-s > 1:
    s +=2
    print(s, end=" ")
```

#### W4A20 : Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không

```python
n=int(input("Nhập vào số nguyên dương n: "))
xac_minh=False
while n > 1:
    n /= 2
    if n % 2 == 0:
        xac_minh=True
if xac_minh == False:
    print("Số không phải dạng 2^k")
else:
    print("Số là dạng 2^k")
```

#### W4A22 : Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n=input("Nhập số nguyên dương n: ")
chu_so_le=0
chu_so_chan=0
for i in range(0, len(n)):
    if n[i] != '0':
        vi_tri_chu_so_dau = i
        break
for s in range(vi_tri_chu_so_dau, len(n)): 
    if int(n[s]) % 2 == 0:
        chu_so_chan += 1
    elif int(n[s]) % 2 != 0:
        chu_so_le += 1
print(f"Số {n} có {chu_so_le} chữ số lẻ và {chu_so_chan} chữ số chẵn")
```

#### W4A24: Nhập vào A, tìm n nhỏ nhất sao cho. 1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
A=float(input("Nhập số thực A: "))
k=0
tong=0
while tong + 1/(k+1) < A:
    k += 1
    tong += 1/k
print(k)
```

#### W4A26
[Fibonaci1]
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A

```python
A = int(input("Nhập số nguyên dương A: "))
f1, f2 = 1, 1
while f1 + f2 <= A:
    f1, f2 = f2, f1 + f2
# Kết quả là f2 (số lớn nhất vừa tính được mà vẫn <= A)
print(f2)
```

#### W4A27
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có
ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng)
Dùng cách: for trên index; KHÔNG dùng hàm split(), len()

```python
chuoi = input("Nhập chuỗi: ")

do_dai = 0
for ky_tu in chuoi:
    do_dai += 1

so_tu = 0
dang_trong_tu = False

for i in range(do_dai):
    if chuoi[i] != " ":
        if dang_trong_tu == False:
            so_tu += 1
            dang_trong_tu = True
    else:
        dang_trong_tu = False

print(f"Số từ trong chuỗi là: {so_tu}")
```

#### W4A28
Nhập vào một chuỗi, hãy in từ đầu tiên trong chuỗi
Dùng cách: for trên index; KHÔNG dùng hàm split()

```python
chuoi = input("Nhập chuỗi: ")

tu_dau_tien = ""
dang_trong_tu = False

for i in range(len(chuoi)):
    if chuoi[i] != " ":
        tu_dau_tien += chuoi[i]
        dang_trong_tu = True
    else:
        if dang_trong_tu == True:
            break

print(f"Từ đầu tiên là: {tu_dau_tien}")
```

#### W4A29 Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành thường và thường thành hoa.

```python
chuoi = input("Nhập chuỗi: ")
chuoi_moi = ""

for i in chuoi:
    if i.isupper():
        chuoi_moi += i.lower()
    elif i.islower():
        chuoi_moi += i.upper()
    else:
        chuoi_moi += i # Giữ nguyên nếu là khoảng trắng hoặc số

print(chuoi_moi)
```
