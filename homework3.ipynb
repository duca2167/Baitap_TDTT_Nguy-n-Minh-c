#### Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím

```python
h = int(input("Nhập độ cao h muốn có của tam giác: "))

for i in range(1, h + 1):
    # Tính số khoảng trắng ở đầu dòng
    khoang_trang_dau = " " * (h - i)
    
    if i == 1:
        # Dòng đầu tiên chỉ có 1 dấu sao
        print(khoang_trang_dau + "*")
    elif i == h:
        # Dòng cuối cùng là đáy tam giác (in toàn bộ dấu sao)
        # Số dấu sao ở đáy tam giác cân cao h luôn là 2*h - 1
        print("*" * (2 * h - 1))
    else:
        # Các dòng ở giữa: Khoảng trắng đầu + * + Khoảng trắng giữa + *
        # Số khoảng trắng giữa = 2*i - 3
        khoang_trang_giua = " " * (2 * i - 3)
        print(khoang_trang_dau + "*" + khoang_trang_giua + "*")
```

#### Bài 15: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số. Làm bằng nhiều cách nhất có thể, ko dùng mảng

```python
# Cách 1: Dùng string
def xet_so_chu_so(n):
    for i in range(0, len(n)):
        if n[i] != '0':
            so_chu_so = len(n) - i
            return so_chu_so
        elif n[i] == '0':
            continue

print("Số đã nhập có số chữ số là: ", xet_so_chu_so(n=input()))
```

#### Bài 16: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

#### Bài 18: Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay không

```python
n = int(input("Nhập vào số nguyên dương n: "))

# Chừng nào n còn chia hết cho 3 và n phải lớn hơn 1
while n > 1 and n % 3 == 0:
    n /= 3

# Nếu cuối cùng n bằng 1 thì nó chính là dạng 3^k
if n == 1:
    print("Yes")
else:
    print("No")
```

#### Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
A = int(input("Nhập số nguyên dương A: "))
f1 = 1  # Số Fibonacci thứ nhất
f2 = 1  # Số Fibonacci thứ hai
f_ke = 1 # Biến lưu số tiếp theo
# Nếu A nhỏ hơn 1 thì không có số nào thỏa mãn (nhưng đề cho A nguyên dương)
if A < 1:
    print("Không có số thỏa mãn")
else:
    while f1 + f2 <= A:
        f_ke = f1 + f2
        # Cập nhật lại hai số trước đó để tính số tiếp theo nữa
        f1 = f2
        f2 = f_ke
    print(f2)
```

#### Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố.

```python
n = input("Nhập số nguyên dương n: ")
chu_so_nguyen_to = 0
for chu_so in n:
    so = int(chu_so)
    # Số nguyên tố phải lớn hơn 1
    if so > 1:
        is_nguyen_to = True 
        # Kiểm tra xem có chia hết cho số nào từ 2 đến chính nó - 1 không
        for i in range(2, so):
            if so % i == 0:
                is_nguyen_to = False # Nếu chia hết thì không phải nguyên tố
                break 
        # Sau khi kiểm tra xong các số i, nếu vẫn là True thì mới đếm
        if is_nguyen_to:
            chu_so_nguyen_to += 1

print(f"Số {n} có {chu_so_nguyen_to} chữ số nguyên tố.")
```
