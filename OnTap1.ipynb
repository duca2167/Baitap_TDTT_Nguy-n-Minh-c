#### 1 Yêu cầu. Cước phí Taxi Thông Minh.
Viết một hàm duy nhất smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính và trả về total (tổng tiền phải trả,
kiểu số nguyên, đơn vị VND).
Các quy tắc cần cài đặt:
1. Giá mở cửa (0–0.8 km):
- Xe 4 chỗ: 11,000 VND
- Xe 7 chỗ: 12,000 VND
2. Giá theo km sau 0.8 km:
- Xe 4 chỗ: từ 0.8→20 km ơnh 13,000 VND/km; trên 20 km tính 10,500 VND/km
- Xe 7 chỗ: từ 0.8→20 km ơnh 14,000 VND/km; trên 20 km tính 12,000 VND/km
3. Phí chờ: 5 phút đầu miễn phí; sau đó, mỗi 5 phút (làm tròn lên theo từng block 5 phút) tính 2,000 VND.
4. Phụ thu ban đêm: nếu start_hour thuộc [22,23,0,1,2,3,4,5], cộng +20% trên tiền tạm tính (mở cửa + quãng đường +
chờ).
5. Điều kiện đặc biệt:
- Nếu km > 30 và không phải ban đêm, áp dụng giảm 10% trên tổng tiền.
- Nếu tổng tiền cuối cùng < 50,000 nhưng km > 5, đặt tổng tiền bằng 50,000 VND (mức tối thiểu).

```python
import math

def smart_taxi_fare(loai_xe, gio_bat_dau, km, phut_cho):
    # Giá mở cửa
    if loai_xe == 4:
        gia_mo_cua = 11000
    elif loai_xe == 7:
        gia_mo_cua = 12000
    else:
        raise ValueError("Loại xe phải là 4 hoặc 7")
    
    # Giá theo km
    if km <= 0.8:
        gia_km = 0
    else:
        km_tinh = km - 0.8
        if loai_xe == 4:
            if km_tinh <= 19.2:
                gia_km = km_tinh * 13000
            else:
                gia_km = 19.2 * 13000 + (km_tinh - 19.2) * 10500
        else:  # 7 chỗ
            if km_tinh <= 19.2:
                gia_km = km_tinh * 14000
            else:
                gia_km = 19.2 * 14000 + (km_tinh - 19.2) * 12000
    
    # Phí chờ
    if phut_cho <= 5:
        phi_cho = 0
    else:
        phut_cho_tinh = phut_cho - 5
        so_block = math.ceil(phut_cho_tinh / 5)
        phi_cho = so_block * 2000
    
    # Tiền tạm tính
    tien_tam_tinh = gia_mo_cua + gia_km + phi_cho
    
    # Phụ thu ban đêm
    ban_dem = gio_bat_dau in [22, 23, 0, 1, 2, 3, 4, 5]
    if ban_dem:
        phu_thu_dem = tien_tam_tinh * 0.2
        tien_tam_tinh += phu_thu_dem
    
    # Tổng tiền
    tong_tien = tien_tam_tinh
    
    # Điều kiện đặc biệt
    if km > 30 and not ban_dem:
        tong_tien *= 0.9
    
    if tong_tien < 50000 and km > 5:
        tong_tien = 50000
    
    return int(tong_tien)

print(smart_taxi_fare(4, 10, 15, 10))  
```

```python
def xet_tong(x):
    if x <= 0:
        return False
    total = 0
    for i in range(1, x + 1):
        total += i
        if total == x:
            return True
        if total > x:
            return False
    return False

# Ví dụ sử dụng
x = int(input("Nhập số nguyên x: "))
print(xet_tong(x))
```

Quy tắc:
- Nếu a == 0 và b != 0: trả về nghiệm duy nhất [-c / b].
- Nếu a == 0 và b == 0: trả về [] (không có nghiệm hữu hạn duy nhất nếu c == 0, hoặc vô nghiệm nếu c != 0).
- Với a != 0, dùng biệt thức d = bb - 4ac:
- d < 0 → []
- d == 0 → [ -b / (2a) ]
- d > 0 → hai nghiệm [r1, r2] với r1 <= r2.

```python
import math
#Hàm giải phương trình bậc hai với ax^2 + bx + c = 0
def giai_pt_bac_hai(a, b, c):
    nghiem = []
    if a == 0:
        if b == 0:
            return "Phương trình vô nghiệm" if c != 0 else "Phương trình vô số nghiệm"
        else:
            return f"Phương trình có một nghiệm: x = {-c / b}"
    
    delta = b**2 - 4*a*c
    if delta < 0:
        return "Phương trình vô nghiệm"
    elif delta == 0:
        x = -b / (2*a)
        return f"Phương trình có nghiệm kép: x = {x}"
    else:
        sqrt_delta = math.sqrt(delta)
        x1 = (-b + sqrt_delta) / (2*a)
        x2 = (-b - sqrt_delta) / (2*a)
    nghiem.append(x1)
    nghiem.append(x2)
    return sorted(nghiem)

#Chương trình chính
a = float(input("Nhập hệ số a: "))
b = float(input("Nhập hệ số b: "))
c = float(input("Nhập hệ số c: "))
print(giai_pt_bac_hai(a, b, c))
```

```python
#Hàm sắp xếp ba số nguyên theo thứ tự không giảm
def sap_xep(a, b, c):
    if a > b:
        a, b = b, a
    if b > c:
        b, c = c, b
    if a > b:
        a, b = b, a
    return a, b, c
#Chương trình chính
a = int(input("Nhập số nguyên a: "))
b = int(input("Nhập số nguyên b: "))
c = int(input("Nhập số nguyên c: "))
print(sap_xep(a, b, c))
```

```python
#Hàm sắp xếp ba số nguyên theo thứ tự không giảm 
def sap_xep(a, b, c):
    so = [a, b, c]
    #Dùng selection sort để sắp xếp
    for i in range(len(so)):
        min_index = i
        for j in range(i + 1, len(so)):
            if so[j] < so[min_index]:
                min_index = j
        so[i], so[min_index] = so[min_index], so[i]
    return so 

#Chương trình chính
a, b, c = list(map(int, input("Nhập ba số nguyên a, b, c: ").split()))
print(sap_xep(a, b, c))
```

#### 14 Viết hàm trả về hệ số nhị thức C(n, k).

```python
def nhi_thuc(n, k):
    if k > n or k < 0:
        return 0
    elif k == 0 or k == n:
        return 1
    else:
        result = 1
        for i in range(k):
            result *= (n - i)
            result /= (i + 1)
        return result 
n = int(input("Nhập số nguyên n: "))
k = int(input("Nhập số nguyên k: "))
print(nhi_thuc(n, k))
```

#### 16. Viết hàm trả về góc ∠ABC theo đơn vị độ khi biết tọa độ các điểm A(x1,y1), B(x2,y2), C(x3,y3).

```python
import math
A=tuple(map(float, input("Nhập tọa độ điểm A: ").split()))
B=tuple(map(float, input("Nhập tọa độ điểm B: ").split()))
C=tuple(map(float, input("Nhập tọa độ điểm C: ").split()))

#Tính các vector BA và BC
# ba_vector = (x1-x2, y1-y2)
# bc_vector = (x3-x2, y3-y2)

#Hàm tính góc
import math

def tinh_goc_ba_bc(A, B, C):
    
    # Vector BA = (xA - xB, yA - yB)
    v1 = (A[0] - B[0], A[1] - B[1])
    # Vector BC = (xC - xB, yC - yB)
    v2 = (C[0] - B[0], C[1] - B[1])

    
    # Công thức: x1*x2 + y1*y2
    tich_vo_huong = v1[0] * v2[0] + v1[1] * v2[1]

    
    dodai_v1 = math.sqrt(v1[0]**2 + v1[1]**2)
    dodai_v2 = math.sqrt(v2[0]**2 + v2[1]**2)

    # cos(alpha) = tich_vo_huong / (dodai_v1 * dodai_v2)
    cos_alpha = tich_vo_huong / (dodai_v1 * dodai_v2)

    # Giới hạn giá trị trong [-1, 1] để tránh lỗi sai số máy tính
    cos_alpha = max(-1.0, min(1.0, cos_alpha))

    # Tính radian và chuyển sang độ
    radian = math.acos(cos_alpha)
    return math.degrees(radian)
print(tinh_goc_ba_bc(A, B, C))

```

- Gợi ý: Cài đặt phép kiểm tra điểm trong đa giác tổng quát (ví dụ: ray casting), đồng thời tính cả trường hợp điểm nằm
trên biên là ở trong.

```python
def cross_product(a, b, c):
    """
    Tính tích có hướng của hai vectơ AB và AC.
    Công thức: (Bx-Ax)*(Cy-Ay) - (By-Ay)*(Cx-Ax)
    """
    return (b[0] - a[0]) * (c[1] - a[1]) - (b[1] - a[1]) * (c[0] - a[0])

def point_in_quadrilateral(quad, p):
    # Duyệt qua 4 cạnh của tứ giác
    n = len(quad)
    dau_duong = False
    dau_am = False
    
    for i in range(n):
        # Lấy hai đỉnh liên tiếp để tạo thành một cạnh
        p1 = quad[i]
        p2 = quad[(i + 1) % n] # Dùng % n để quay lại đỉnh đầu tiên khi ở đỉnh cuối
        
        # Tính tích có hướng của cạnh (p1, p2) và điểm p
        cp = cross_product(p1, p2, p)
        
        if cp > 0:
            dau_duong = True
        elif cp < 0:
            dau_am = True
            
        # Nếu tích có hướng có cả dương và âm, điểm nằm ngoài
        if dau_duong and dau_am:
            return False
            
    # Nếu chỉ toàn dương hoặc toàn âm (hoặc có bằng 0 - nằm trên biên)
    return True

quad_test = [(0,0), (4,0), (4,3), (0,3)]
p_test = (2,1)
print(f"Kết quả kiểm tra: {point_in_quadrilateral(quad_test, p_test)}") 
```

#### 26 Yêu cầu: Viết hàm are_collinear(x1, y1, x2, y2, x3, y3) trả về True nếu ba điểm thẳng hàng, ngược lại trả về False.
- Đầu vào: ba điểm x1, y1, x2, y2, x3, y3 biểu diễn tọa độ của ba điểm.
- Đầu ra: giá trị boolean.
- Gợi ý: Dùng phép kiểm tra diện tích (tích có hướng) và một sai số nhỏ khi so sánh số thực. Các điểm nằm trên cùng một
đường thẳng được xem là thẳng hàng (kể cả khi có điểm trùng nhau).

```python
def are_collinear(x1, y1, x2, y2, x3, y3):
    # vectơ u = (x2 - x1, y2 - y1)
    # vectơ v = (x3 - x1, y3 - y1)
    # kiểm tra định thức u x v = 0 để xác định ba điểm có thẳng hàng hay không
    eps = 1e-9
    dx1 = x2 - x1
    dy1 = y2 - y1
    dx2 = x3 - x1
    dy2 = y3 - y1

    if abs(dx1) < eps and abs(dy1) < eps and abs(dx2) < eps and abs(dy2) < eps:
        return "Ba điểm trùng nhau"
    if abs(dx1 * dy2 - dy1 * dx2) < eps:
        return "Ba điểm thẳng hàng"
    return "Ba điểm không thẳng hàng" 
x1, y1 =map(float, input().split())
x2, y2 =map(float, input().split())
x3, y3 =map(float, input().split())
print(are_collinear(x1, y1, x2, y2, x3, y3))
```
