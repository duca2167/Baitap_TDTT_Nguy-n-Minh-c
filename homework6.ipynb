#### W6A16
Nhậ̣p vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma
trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải

```python
n = int(input("Nhập số hàng:"))
m = int(input("Nhập số cột: "))
matrix = []
for i in range(n):
    print(f"Nhập các giá trị của hàng {i+1}")
    hang = list(map(int, input().split()))
    matrix.append(hang)
for i in range(n):
    for j in range(m):
        print(f"{matrix[i][j]:4d}", end="")
    print()
```

#### W6A17
Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và
đường chéo phụ của mảng

```python
n = int(input("Nhập kích cỡ nxn của ma trận vuông: "))
matrix = []
for i in range(n):
    print(f"Nhập các giá trị của hàng {i+1}: ")
    hang = list(map(int, input().split()))
    matrix.append(hang)

def duong_cheo_chinh(matrix, n):
    for i in range(n):
        print("    " * i, end="") 
        print(f"{matrix[i][i]:4d}")

def duong_cheo_phu(matrix, n):
    for i in range(n):
        print("    " * (n - 1 - i), end="")
        print(f"{matrix[i][n - 1 - i]:4d}")

print("Đường chéo chính là:")
duong_cheo_chinh(matrix, n) 

print("Đường chéo phụ là:")
duong_cheo_phu(matrix, n)   
```
