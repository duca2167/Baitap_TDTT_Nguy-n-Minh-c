#### W7A1
Viết hàm tìm kiếm nhị phân
trên một danh sách đã
được sắp xếp tăng dần.
Trả về chỉ số của phần tử
hoặc -1 nếu không tìm
thấy. mẫu hàm: def
binary_search(arr, left,
right, target): 

```python
def binary_search(arr, left, right, target):
    if right >= left:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] > target:
            return binary_search(arr, left, mid - 1, target)
        else:
            return binary_search(arr, mid + 1, right, target)
    else:
        return -1
arr = list(map(int, input().split()))
target = int(input())
result = binary_search(arr, 0, len(arr) - 1, target)
print(arr)
print(result)
```

#### W7A2
Cho danh sách arr và một
số nguyên x, đếm xem x
xuất hiện bao nhiêu lần
trong danh sách. def
count_occurrences(arr, x) -> int:

```python
def count_occurrences(arr, x):
    count = 0
    for num in arr:
        if num == x:
            count += 1
    return count
arr = list(map(int, input().split()))
x = int(input())
occurrences = count_occurrences(arr, x)
print(occurrences)
```

#### W7A3
Viết hàm thực hiện thuật
toán bubble sort, Selection
Sort để sắp xếp danh sách
theo thứ tự tăng dần.

```python
#Bubble Sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
arr = list(map(int, input().split()))
bubble_sort(arr)
print(arr)
```

```python
#Selection Sort
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
arr = list(map(int, input().split()))
selection_sort(arr)
print(arr)
```

#### W7A4
Viết hàm tìm phần tử có số
lần xuất hiện nhiều nhất
trong danh sách. Nếu có
nhiều phần tử cùng số lần,
trả về phần tử xuất hiện
trước.

```python
def phan_tu_nhieu_nhat(arr):
    dict_doc_nhat = []
    for i in arr:
        if i not in dict_doc_nhat:
            dict_doc_nhat.append(i)
    so_lan_max = 0
    so_nhieu_nhat = dict_doc_nhat[0]
    for i in dict_doc_nhat:
        so_lan = 0
        for j in arr:
            if i == j:
                so_lan += 1
        if so_lan > so_lan_max:
            so_lan_max = so_lan
            so_nhieu_nhat = i
    return so_nhieu_nhat, so_lan_max
arr = list(map(int, input().split()))
result = phan_tu_nhieu_nhat(arr)
print(f"{result[0]} xuat hien nhieu nhat, som nhat, {result[1]} lan")
```

phẳng danh sách

#### W7A7
Cho một danh sách các số
nguyên nums, hãy tìm độ
dài của dãy con tăng
nghiêm ngặt dài nhất
(không nhất thiết phải liền
kề).

```python
def do_dai_tang_chuan(arr):
    if not arr:
        return 0
    
    n = len(arr)
    f = [1] * n
    
    for i in range(1, n):
        for j in range(i):
            if arr[i] > arr[j]:
                if f[i] < f[j] + 1:
                    f[i] = f[j] + 1
                    
    return max(f)

arr = list(map(int, input().split()))
print(do_dai_tang_chuan(arr))
```

#### W7A8
Bạn được cung cấp: một
danh sách các khoảng thời
gian interval = [[l1, r1], [l2,
r2], ...] và một danh sách
các truy vấn queries = [q1,
q2, ...]
Với mỗi truy vấn q, hãy tìm
độ dài nhỏ nhất của
khoảng thời gian [li, ri] sao
cho li ≤ q ≤ ri.
Nếu không tồn tại khoảng
thời gian nào như vậy, hãy
trả về -1 cho truy vấn đó.
solve in < O(m.n), hint:
binary search in the
queries list

```python
intervals = eval(input())
queries = eval(input())

for i in range(len(intervals)):
    for j in range(i + 1, len(intervals)):
        len_i = intervals[i][1] - intervals[i][0] + 1
        len_j = intervals[j][1] - intervals[j][0] + 1
        if len_i > len_j:
            intervals[i], intervals[j] = intervals[j], intervals[i]

results = []


for q in queries:
    found = -1
    for l, r in intervals:
        if l <= q <= r:
            found = r - l + 1
            break 
    results.append(found)

print(*results)
```

#### W7A9
Cho một danh sách các số
nums, với mỗi phần tử
nums[i], hãy đếm xem có
bao nhiêu phần tử bên phải
nhỏ hơn nums[i].
Trả về danh sách các số
đếm.
Efficient solution: using
heap or binary search tree

```python
def bigger_on_the_right(arr):
    result = []
    for i in range(len(arr)):
        count = 0
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[i]:
                count += 1
        result.append(count)
    return result    
arr = list(map(int, input().split()))
print(bigger_on_the_right(arr))
```

#### W7A10
Cho một danh sách các số
nguyên nums và một số
nguyên m, hãy tìm tổng
mảng con lớn nhất theo
modulo m.
Tức là, với bất kỳ mảng
con nums[i:j] nào, hãy tính
sum(nums[i:j]) % m và trả
về kết quả lớn nhất có thể.
Gợi ý:
Duy trì tổng tiền tố (prefix
sum) mod m và sử dụng
tìm kiếm nhị phân để tìm
tổng tiền tố nhỏ nhất chỉ
lớn hơn tổng hiện tại một
chút (để tối đa hóa hiệu số
bao quanh).

```python
nums = list(map(int, input().split()))
m = int(input())
max_mod_sum = 0
for i in range(len(nums)):
    current_sum = 0
    for j in range(i, len(nums)):
        current_sum = (current_sum + nums[j]) % m
        if current_sum > max_mod_sum:
            max_mod_sum = current_sum
print(max_mod_sum)
```

#### W7A13
Dãy hài hòa là một dãy
thỏa mãn điều kiện giá trị
lớn nhất và giá trị nhỏ nhất
của dãy khác nhau đúng
bằng \( 1 \).
Viết hàm trả về độ dài của
dãy hài hòa con dài nhất có
thể có trong mảng giá trị
nguyên \( arr \) có \( n \)
phần tử
Ví dụ:
Input: 1 3 2 2 5 2 3 7
Output: 5 - Dãy hài hòa con
dài nhất là [3,2,2,2,3]

```python
def do_dai_day_hai_hoa_dai_nhat(arr):
    count = {}
    for num in arr:
        if num in count:
            count[num] += 1
        else:
            count[num] = 1
            
    max_length = 0
    
    for x in count:
        if (x + 1) in count:
            current_length = count[x] + count[x + 1]
            if current_length > max_length:
                max_length = current_length
                
    return max_length

arr = list(map(int, input().split()))
print(do_dai_day_hai_hoa_dai_nhat(arr))
```
