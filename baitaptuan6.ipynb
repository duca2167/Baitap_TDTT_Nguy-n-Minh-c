#### W6A3
Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple
sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp
xếp.)

```python
def rotate_left(lst, k):
    k %= len(lst)
    return lst[-k:] + lst[:-k]

# Chương trình chính
dong_so_nguyen = tuple(map(int, input().split()))
k = int(input("Nhập dòng kế: "))
print(rotate_left(dong_so_nguyen, k))
```

#### W6A4
Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo
dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key
lặp lại, thêm vào list của key đó

```python
input_data = input("Nhập các cặp key:value: ").split()
data_map = {}
for item in input_data:
    key, value = item.split(':')
    if key not in data_map:
        data_map[key] = [value] 
    else:
        data_map[key].append(value) 

print(data_map)
```

#### W6A5
Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ...,
'zeros': ...} (đếm số >0, <0, =0)

```python
lst_so_nguyen = list(map(int, input().split()))
dem_so = {
    'positives': 0,
    'negatives': 0,
    'zeros': 0
}

for i in lst_so_nguyen:
    if i > 0:
        dem_so['positives'] += 1
    elif i < 0:
        dem_so['negatives'] += 1
    else:
        dem_so['zeros'] += 1

print(dem_so)
```

#### W6A9
Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một
dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ
được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới
theo thứ tự từ điển tăng dần của key.

```python
def nhap_dict(prompt):
    print(prompt)
    data = input("Nhập các cặp key:value: ").split()
    d = {}
    for item in data:
        k, v = item.split(':')
        d[k] = int(v) # Chuyển value sang kiểu số để cộng được
    return d

# Nhập 2 dictionary
dict1 = nhap_dict("--- Nhập Dictionary 1 ---")
dict2 = nhap_dict("--- Nhập Dictionary 2 ---")

dict_tong = dict(dict1) # Lấy toàn bộ dữ liệu từ dict1 sang trước

for key, value in dict2.items():
    if key in dict_tong:
        dict_tong[key] += value 
    else:
        dict_tong[key] = value  
dict_sap_xep = {}
for key in sorted(dict_tong):
    dict_sap_xep[key] = dict_tong[key]


print(" Dictionary sau khi gộp và sắp xếp:")
print(dict_sap_xep)
```

dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là
một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh
sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được
sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple.

```python
lst = list(map(int, input().split()))
k = int(input("Nhập số nguyên k: "))
lst_da_sap_xep = []
for i in range(len(lst)-1):
    for j in range (i+1, len(lst)):
        if lst[i] + lst[j] == k:
            tup = (lst[i], lst[j])
            lst_da_sap_xep.append(tup)
print(set(lst_da_sap_xep))

```

#### W6A13
[Đảo ngược key và value của dictionary]
Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo
một dictionary mới bằng cách đảo ngược key và value của dictionary ban
đầu. In ra dictionary mới.

```python
def nhap_dict():
    data = input("Nhập các cặp key:value: ").split()
    d = {}
    for item in data:
        k, v = item.split(':')
        d[k] = v 
    return d
dict_ban_dau = nhap_dict()
dict_moi = {}
for key in dict_ban_dau:
    dict_moi[dict_ban_dau[key]] = key

print(dict_moi)
```

#### W6A14 [Giao của hai danh sách]
Nhập vào hai dòng, mỗi dòng là một danh sách các số nguyên. Hãy tìm các
phần tử chung (giao) của hai danh sách này và in ra một list mới chứa các
phần tử chung đó, không trùng lặp.

```python
lst1 = list(map(int, input().split()))
lst2 = list(map(int, input().split()))
giao = []
for so in lst1:
    if so in lst2:
        if so not in giao:
            giao.append(so)
        else:
            continue 
print(giao)
```

#### W6A15
[Lọc dictionary theo giá trị]
Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số
nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp
key-value từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary
mới.

```python
def nhap_dict():
    print("Hãy nhập vào một dict với key là chuỗi và value là số nguyên, cách bằng dấu cách: ")
    data = input().split()
    d = {}
    for item in data:
        k, v=item.split(":")
        d[k] = v
    return d
dict_ban_dau = nhap_dict()
so_k = int(input("Nhập số nguyên k bất kỳ: "))
dict_moi = {}
for k, v in dict_ban_dau.items():
    if int(v) > so_k:
        dict_moi[k] = int(v)

print(dict_moi)
```
