#### W9A16
Cho hai số nguyên dương, n và m (0 < n, m <= 20 ). .Bạn hãy in ra tất cả các chuỗi nhị phân có
n số 0 và m số 1. In theo thứ tự tăng dần trong từ điển

```python
def doi_so_nhi_phan(str_hien_tai, n, m):
    if n == 0 and m == 0:
        print(str_hien_tai, end = " ")
        return
    
    if n > 0:
        doi_so_nhi_phan(str_hien_tai + "0", n - 1, m)
        
    if m > 0:
        doi_so_nhi_phan(str_hien_tai + "1", n, m - 1)

n = int(input())
m = int(input())

doi_so_nhi_phan("", n, m)
```

#### W9A17 Viết hàm tìm dãy con tăng nghiêm ngặt dài nhất trong dãy đầu vào từ bàn phím

```python
def day_con_tang_nghiem_ngat(arr):
    if not arr:
        return []
    else:
        n = len(arr)
        cac_day_con = [[] for _ in range(n)]
        for i in range(n):
            cac_day_con[i].append(arr[i])
        for i in range(1, n):
            for j in range(i):
                if arr[j] < arr[i]:
                    if len(cac_day_con[j]) + 1 > len(cac_day_con[i]):
                        cac_day_con[i] = cac_day_con[j] + [arr[i]]
        day_dai_nhat = cac_day_con[0]
        for i in range(1, n):
            if len(cac_day_con[i]) > len(day_dai_nhat):
                day_dai_nhat = cac_day_con[i]
        return day_dai_nhat
arr = list(map(int, input().split()))
ket_qua = day_con_tang_nghiem_ngat(arr)
print("Dãy con tăng nghiêm ngặt dài nhất là:", ket_qua)
```

#### W9A18
Cho dãy N phần tử số nguyên, hãy đếm số cặp có vị trí khác nhau và có tổng bằng k (số nguyên
dương cho vào từ bàn phím)

```python
arr = list(map(int, input().split()))
k = int(input())
n = len(arr)
count = 0

for i in range(n - 1):
    for j in range(i + 1, n):
        if arr[i] + arr[j] == k:
            count += 1

print(count)
```

Tìm phần tử lớn thứ k trong dãy n phần tử bằng thuật toán tốt nhất có thể – Tránh dùng sắp xếp
(dùng phân hoạch của quicksort)

```python
arr = list(map(int, input().split()))
k = int(input())
arr.sort()
print(f"Phần tử lớn thứ {k} là {arr[-k]}")

```

```python
def quickselect(arr, k):
    pivot = arr[0]
    left = []
    mid = []
    right = []
    
    for x in arr:
        if x > pivot:
            left.append(x)
        elif x == pivot:
            mid.append(x)
        else:
            right.append(x)
            
    if k <= len(left):
        return quickselect(left, k)
    elif k <= len(left) + len(mid):
        return pivot
    else:
        return quickselect(right, k - len(left) - len(mid))

n = int(input())
arr = list(map(int, input().split()))
k = int(input())

print(quickselect(arr, k))
```

#### W9A22
Hai chuỗi là anagram nếu chúng có cùng ký tự và số lần xuất hiện. viết CT kiểm tra xem 2 chuỗi
đầu vào có phải là anagram của nhau không

```python
s1 = input()
s2 = input()

if len(s1) != len(s2):
    print("False")
else:
    counts = {}
    for char in s1:
        if char in counts:
            counts[char] += 1
        else:
            counts[char] = 1
    
    for char in s2:
        if char in counts:
            counts[char] -= 1
        else:
            print("False")
            exit()
            
    is_anagram = True
    for count in counts.values():
        if count != 0:
            is_anagram = False
            break
            
    print(is_anagram)
```

#### W9A23
Cho số nguyên rất lớn (string), và một số nguyên dương k. Tìm số có k chữ số lớn nhất có thể
từ số ban đầu bằng cách xóa bớt đi một số chữ số nào đó mà không thay đổi vị trí của bất kì
chữ số nào. (VD: 78934340999, k = 7 => 9440999 là số lớn nhất có 7 chữ số cần tìm)

```python
s = input()
k = int(input())
res = ""
start = 0
n = len(s)

for i in range(k, 0, -1):
    end = n - i
    max_digit = '0'
    max_idx = start
    
    for j in range(start, end + 1):
        if s[j] > max_digit:
            max_digit = s[j]
            max_idx = j
    
    res += max_digit
    start = max_idx + 1

print(res)
```

#### W9A25
Giả sử mật khẩu mạnh là một chuỗi kí tự dài ít nhất 8 kí tự gồm ít nhất 1 chữ cái thường, 1 chữ
cái hoa, 1 chữ số và 1 trong các kí tự đặc biệt !, @, #, $. Nhập vào một chuỗi kĩ tự, kiểm tra xem
nó có phải là mật khẩu mạnh không.

```python
password = input()
co_thuong = False
co_hoa = False
co_so = False
co_ky_tu_dac_biet = False
for ky_tu in password:
    if ky_tu.islower():
        co_thuong = True
    elif ky_tu.isupper():
        co_hoa = True
    elif ky_tu.isdigit():
        co_so = True
    elif ky_tu in ['!', '@', '#', '$']:
        co_ky_tu_dac_biet = True
if co_thuong and co_hoa and co_so and co_ky_tu_dac_biet:
    print("Strong password")
else:
    print("Weak password")
```
