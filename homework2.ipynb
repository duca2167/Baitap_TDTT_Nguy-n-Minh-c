#### Bài 1: Nhập vào 2 cạnh của một hình chữ nhật. In ra màn hình diện tích và chu vi của nó

#### Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là độ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân,Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn hình.

```python
a=float(input("Nhập số a:"))
b=float(input("Nhập số b:"))
c=float(input("Nhập số c:"))
if a+b<c or a+c<b or b+c<a:
    print("Khong la ba canh của mot tam giac")
else:
    if a==b and b==c:
        print("Tam giac deu")
    elif a==b or b==c or a==c:
        print("Tam giac can")
    elif a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2:
        print("Tam giac vuong")
    else:
        print("Tam giac thuong")
    p=(a+b+c)/2
    s=math.sqrt(p*(p-a)*(p-b)*(p-c))
    print("Diện tích tam giác là:", round(s, 2))
    print("Chu vi tam giác là:", round(a+b+c, 2))
```

#### Bài 5: Viết chương trình giải phương trình bậc 2 (ax^2 + bx + c = 0)

```python
print("*******GIẢI PHƯƠNG TRÌNH BẬC HAI ax^2+bx+c=0*******")
a=float(input("Nhập hệ số a:"))
b=float(input("Nhập hệ số b:"))
c=float(input("Nhập hệ số c:"))
if a==0:
    if b==0:
        if c==0:
            print("Phương trình có vô số nghiệm")
        else:
            print("Phương trình vô nghiệm")
    else:
        x=-c/b
        print("Nghiệm của phương trình là:", round(x, 2))
else:
    delta=b**2-4*a*c
    if delta<0:
        print("Phương trình vô nghiệm")
    elif delta==0:
        x=-b/(2*a)
        print("Phương trình có nghiệm kép:", round(x, 2))
    else:
        x1=(-b+math.sqrt(delta))/(2*a)
        x2=(-b-math.sqrt(delta))/(2*a)
        print("Phương trình có hai nghiệm phân biệt:")
        print("Nghiệm 1:", round(x1, 2))
        print("Nghiệm 2:", round(x2, 2))
        
```

#### Bài 8: Giải hệ phương trình tuyến tính: ax + by = m; cx + dy = n

```python
print("*****GIẢI HỆ PHƯƠNG TRÌNH TUYẾN TÍNH 2 ẨN*****")
print("Hệ phương trình có dạng:")
print(f"ax + by = m")
print(f"cx + dy = n")
a=float(input("Nhập hệ số a:"))
b=float(input("Nhập hệ số b:"))
c=float(input("Nhập hệ số c:"))
d=float(input("Nhập hệ số d:"))
m=float(input("Nhập hệ số m:"))
n=float(input("Nhập hệ số n:"))
det=a*d-c*b
if det==0:
    if a/c == b/d == m/n:
        print("Hệ phương trình có vô số nghiệm")
    else:
        print("Hệ phương trình vô nghiệm")
else:
    x=(m*d-n*b)/det
    y=(a*n-m*c)/det
    print("Nghiệm của hệ phương trình là:")
    print("x =", round(x, 2))
    print("y =", round(y, 2))
```

#### Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
x, y = map(float, input("Nhập tọa độ tâm đường tròn (x, y): ").split())
r=float(input("Nhập bán kính r: "))
a, b = map(float, input("Nhập tọa độ điểm (a, b): ").split())
distance = math.sqrt((a - x)**2 + (b - y)**2)
if distance < r:
    print("Điểm nằm trong đường tròn")
elif distance == r:
    print("Điểm nằm trên đường tròn")
else:
    print("Điểm nằm ngoài đường tròn")
```
