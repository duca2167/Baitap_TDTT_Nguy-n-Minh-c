Bài 2B: Viết hàm zigzag_numbers(n) trả về một chuỗi biểu diễn ma trận n x n chứa các số từ 1 đến n^2. Các hàng cách nhau bởi ký tự xuống 
dòng. Với chỉ số hàng bắt đầu từ 0: hàng chẵn đi từ trái sang phải, hàng lẻ đi từ phải sang trái. Các số trong cùng một hàng cách nhau bởi 
một dấu cách. 
n = 2
1 2
4 3
n= 4
1 2 3 4
8 7 6 5
9 10 11 12
16 15 14 13

```python
def zigzag_numbers(n):
    rows = []
    for i in range(n):
        start = i * n + 1
        end = (i + 1) * n + 1
        
        row_numbers = list(range(start, end))
        
        if i % 2 != 0:
            row_numbers.reverse()
            
        
        row_str = " ".join(map(str, row_numbers))
        rows.append(row_str)
        
    
    return "\n".join(rows)

print(zigzag_numbers(4))
```

```python
def zigzag_numbers(n):
    lst = []
    x = 1
    for i in range(n):
        lst.append([])
        for j in range(n):
            lst[i].append(x)
            x += 1
    for i in range(len(lst)):
        if i % 2 == 1:
            lst[i].reverse()
    return lst
n = int(input())
print(zigzag_numbers(n))
```

Bài 3: Trả về cột (chỉ số bắt đầu từ 0) có tổng cột lớn nhất trong ma trận n x m. 
Viết hàm column_with_max_sum(mat) trả về chỉ số bắt đầu từ 0 của cột có tổng lớn nhất. 
Nếu có nhiều cột cùng tổng lớn nhất, trả về cột có chỉ số nhỏ nhất. - Dùng vòng lặp để tính tổng từng cột. - Không dùng tuple hoặc dictionary. 
n = 2
m = 3
mat = [[1, 3, 5], 
       [6, 2, -9]]


```python
n, m = map(int, input().split())
matrix = []
for _ in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)

def column_with_max_sum(matrix):
    if not matrix or not matrix[0]:
        return 0
        
    max_sum = 0
    max_column = 0
    
    for i in range(m):
        current_sum = 0
    
        for j in range(n):
            current_sum += matrix[j][i]
            
        
        if current_sum > max_sum:
            max_sum = current_sum
            max_column = i+1
            
    return max_column

for row in matrix:
    print(row)
print(column_with_max_sum(matrix))
```

Bài 4: Trả về hàng (chỉ số bắt đầu từ 0) có tổng hàng nhỏ nhất trong ma trận n x m. 
Viết hàm row_with_min_sum(mat) trả về chỉ số bắt đầu từ 0 của hàng có tổng nhỏ nhất. 
Nếu có nhiều hàng cùng tổng nhỏ nhất, trả về hàng có chỉ số lớn nhất. - Dùng vòng lặp để tính tổng từng hàng. -. 
n = 2
m = 3
mat = [[1, 3, 5], 
       [6, 2, -9]]


```python
n, m = map(int, input().split())
matrix = []
for _ in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)
def row_with_min_sum(matrix):
    if not matrix or not matrix[0]:
        return 0
    else:
        min_tong = 99
        min_row = 0
        for i in range(n):
            tong = 0
            for j in range(m):
                tong += matrix[i][j]
            if tong <= min_tong:
                min_tong = tong
                min_row = i + 1
        return min_row
for row in matrix:
    print(row)
print(row_with_min_sum(matrix))
```

Bài 5: 
Cho ma trận vuông cấp n. In ra đường chéo chính và đường chéo phụ của ma trận


```python
n = int(input())
matrix = []
for _ in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)

def duong_cheo_chinh(matrix):
    for i in range(n):
        print(matrix[i][i], end=" ")
    print() 

def duong_cheo_phu(matrix):
    for i in range(n):
        print(matrix[i][n - 1 - i], end=" ")
    print() 

# Gọi hàm thực thi
duong_cheo_chinh(matrix)
duong_cheo_phu(matrix)
```

Bài 6: Cho ma trận nxm số nguyên, thêm vào một cột tại vị trí k.

```python
n = int(input())
m = int(input())
k = int(input())
col_data = list(map(int, input("Cột muốn thêm vào: ").split ()))
matrix = []
for _ in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)

def them_cot_tai_k(matrix, col_data, k):
    n = len(matrix)
    m = len(matrix[0])
    if k < 0:
        k = 0
    elif k > m:
        k = m
    new_matrix = []
    for i in range(n):
        new_row = []
        for j in range(m+1):
            if j < k:
                new_row.append(matrix[i][j])
            elif j == k:
                new_row.append(col_data[i])
            else:
                new_row.append(matrix[i][j-1])
        new_matrix.append(new_row)
    return new_matrix
result = them_cot_tai_k(matrix, col_data, k)
for row in result:
    print(row)
    
```
